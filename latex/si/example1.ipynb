```python
import poolparty as pp

# Mark changes by changing case
pp.set_default('mark_changes',True)

# Define sequences (lowercase wild-type so changes appear uppercase)
wt_seq = 'ATCGATCGATATCGATCGAT'         # 10 nt wild-type
adapter_5_seq = 'GAAC'                  # 3 nt 5' adapter
adapter_3_iupac = 'TGGA.NNNNN.CCAG'     # 3 + 5N barcode + 3

# Create pool of randomly mutagenized wt_seq sequences
mut_pool = pp.mutagenize(
    wt_seq, 
    mutation_rate=0.10, 
    op_name='op1')

# Create pool of 3' adaptors with random barcodes
adapter_3_pool = pp.from_iupac_motif(
    adapter_3_iupac, 
    op_name='op2')

# Join into final oligo pool
oligo_pool = pp.join(
    [adapter_5_seq, mut_pool, adapter_3_pool], 
    spacer_str=' ',
    op_name='op3')

# Sample 5 sequences
df = oligo_pool.generate_seqs(num_seqs=5, seed=42)

print(df.columns)
cols = ['seq','']

print(df[['seq']])
```
