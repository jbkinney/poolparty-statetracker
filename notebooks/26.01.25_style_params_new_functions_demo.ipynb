```python
import poolparty as pp; pp.init()

# Demo of style parameter for various functions

# 1. from_iupac with style
pool1 = pp.from_iupac('RN', style='red').named('iupac_red')
print("from_iupac with style='red':")
pool1.print_library()

# 2. from_motif with style
import pandas as pd
prob_df = pd.DataFrame({'A': [0.5], 'C': [0.3], 'G': [0.1], 'T': [0.1]})
pool2 = pp.from_motif(prob_df, style='blue').named('motif_blue')
print("\nfrom_motif with style='blue':")
pool2.print_library(num_seqs=3, seed=42)

# 3. shuffle_seq with style (renamed from style_shuffle)
pool3 = pp.from_seq('ACGTACGT').shuffle_seq(region=[2, 6], style='green').named('shuffle_green')
print("\nshuffle_seq with style='green':")
pool3.print_library(num_seqs=3, seed=42)

# 4. from_fasta with style (would need actual FASTA file, showing pattern)
# pool4 = pp.from_fasta('file.fa', 'chr1', 0, 10, style='cyan')

# 5. rc with style
pool5 = pp.from_seq('ACGT').rc(style='magenta').named('rc_magenta')
print("\nrc with style='magenta':")
pool5.print_library()

# 6. lower with style
pool6 = pp.from_seq('ACGT').lower(style='yellow').named('lower_yellow')
print("\nlower with style='yellow':")
pool6.print_library()

# 7. upper with style
pool7 = pp.from_seq('acgt').upper(style='cyan').named('upper_cyan')
print("\nupper with style='cyan':")
pool7.print_library()

# 8. swapcase with style
pool8 = pp.from_seq('AcGt').swapcase(style='red bold').named('swapcase_red_bold')
print("\nswapcase with style='red bold':")
pool8.print_library()

# 9. slice_seq with style
pool9 = pp.slice_seq(pp.from_seq('ACGTACGT'), slice(2, 6), style='blue').named('slice_blue')
print("\nslice_seq with style='blue':")
pool9.print_library()

# 10. join with style
pool10 = pp.join([pp.from_seq('ACGT'), pp.from_seq('TTTT')], spacer_str='-', style='green').named('join_green')
print("\njoin with style='green':")
pool10.print_library()
```
