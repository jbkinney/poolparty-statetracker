```python
import poolparty as pp; pp.init()


all_kmers = pp.get_kmers(6, mode="sequential").named('all_kmers')
all_kmers.print_library(num_seqs=10)

def balanced_gc(seq):
    gc = (seq.count('G') + seq.count('C')) / len(seq)
    return 0.2 <= gc <= 0.8
filtered = all_kmers.filter(balanced_gc).named('filtered')
filtered.print_library(num_seqs=10)

barcodes = filtered.materialize(num_cycles=10, seed=42).named('barcodes')
barcodes.print_library(num_seqs=10)

mutated = barcodes.mutagenize(num_mutations=1, mode="random", style='red bold').named("mutated_barcodes")
mutated.print_library(num_seqs=5, seed=1)
```
