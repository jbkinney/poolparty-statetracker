```python
import poolparty as pp; pp.init()

# Create a pool with sequences containing markers and mixed case
pool = pp.from_seqs([
    'AAA<M>GGGcccTTT</M>AAA',   # Marker region with lowercase
    'TTT<M>ACG-.-ACG</M>TTT',   # Marker region with gaps
], mode='sequential').named('seqs')

# === Global highlights using add_highlight() ===
print("=== Global highlights (set via add_highlight) ===")
pp.add_highlight('red', which='upper')     # Uppercase letters
pp.add_highlight('cyan', which='lower')    # Lowercase letters  
pp.add_highlight('yellow', which='gap')    # Gap characters

# print_library() automatically uses global highlights when none specified
pool.print_library()

# Clear and set new global highlights
pp.clear_highlights()
pp.add_highlight('bold', region='M', which='tags')
pp.add_highlight('green', region='M', which='contents')
print("=== New global highlights (tags bold, contents green) ===")
pool.print_library()
pool.print_library(show_highlights=False)

# Can still override with explicit highlights parameter
pp.clear_highlights()
print("=== Explicit highlights (overrides global) ===")
pool.print_library(highlights=[pp.Highlighter('magenta', region='M', regex=r'ACG')])
pool.print_library(highlights=[pp.Highlighter('cyan', region='M', regex=r'ACG')])

```
