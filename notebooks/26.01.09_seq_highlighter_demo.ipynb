```python
import poolparty as pp; pp.init()
pp.add_highlight(which='tag', color='green')
pp.add_highlight(region='cre', style='yellow')
pp.add_highlight(region='cre', which='lower', style='cyan')

# Overlapping highlights: later-added highlight takes precedence for foreground colors
pool = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGCAGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')\
    .mutagenize('cre', mutation_rate=0.1, mode='hybrid', num_hybrid_states=3, seq_name_prefix='mut_')

# Yellow background, then cyan for lowercase (mutations) - cyan wins on overlapping chars
pp.clear_highlights()
print("Yellow + Cyan (cyan wins):")
pool.print_library()

# Yellow background, then red for lowercase (mutations) - red wins on overlapping chars
pp.clear_highlights()
pp.add_highlight(region='cre', style='yellow')
pp.add_highlight(region='cre', which='lower', style='red')
print("Yellow + Red (red wins):")
pool.print_library()

# Additive styles: bold + underline + color all combine
pp.clear_highlights()
pp.add_highlight(region='cre', style='bold')
pp.add_highlight(region='cre', style='underline')
pp.add_highlight(region='cre', which='lower', style='magenta')
print("Bold + Underline + Magenta (all combine):")
pool.print_library()

```
