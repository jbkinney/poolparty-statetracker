```python
# Sequential enumeration: 'RN' -> R(A/G) x N(A/C/G/T) = 8 sequences
import poolparty as pp

with pp.Party() as party:
    pool = pp.from_iupac_motif('RN', mode='sequential').named('iupac')

print(f"IUPAC pattern 'RN' has {pool.num_states} states")
df = pool.generate_seqs(num_cycles=1)
df
```

```python
# Random mode: sample from IUPAC pattern
with pp.Party() as party:
    pool = pp.from_iupac_motif('NNNNGYNNNN', mode='random').named('random_6mer')

df = pool.generate_seqs(num_seqs=10, seed=42)
df
```

```python
# Chaining: join fixed flanks with IUPAC variable region
with pp.Party() as party:
    left = pp.from_seq('GAATTC').named('left')   # EcoRI site
    var = pp.from_iupac_motif('NNNN', mode='sequential').named('var')
    right = pp.from_seq('GGATCC').named('right') # BamHI site
    construct = pp.join([left, var, right]).named('construct')

print(f"Total constructs: {construct.num_states}")
df = construct.generate_seqs(num_cycles=1)
df[['seq', 'var.seq']]
```
