```python
import poolparty as pp 
pp.init()
pp.add_highlight(which='tags', style='gray')
pp.add_highlight(region='site_a', style='lightskyblue')
pp.add_highlight(region='site_b', style='chocolate')
#pp.add_highlight(which='lower gap', style='bold yellow')
```

```python
pp.clear_pools()

# 1. Create template Pool
pool = pp.from_seq('TCCCGACT<site_a>GGAAGCG</site_a>GCAGTGAGCACACAGGAAA<site_b>ATTACG</site_b>AGATCGGA')
pool.print_library()
```

```python
pool.insert_kmers('site_a', length=6, mode='hybrid', num_hybrid_states=5)\
    .insert_kmers('site_b', length=6, mode='hybrid', num_hybrid_states=5)\
    .print_library()
```

```python
site_a = pp.get_kmers(length=6, mode='hybrid', num_hybrid_states=5)
site_b = pp.get_kmers(length=6, mode='hybrid', num_hybrid_states=5)

combo = pp.join([site_a, site_b], spacer_str=' ')
combo.print_library()

```
