```python
import poolparty as pp

with pp.Party():
    pool = pp.marker_multiscan(
        'ACGTACACCATGAT',
        markers=['m0', 'm1', 'm2'],
        num_insertions = 5,
        insertion_mode='unordered_with_replacement',
        mode='random',
        op_name='ms',
    ).named('demo')

    df = pool.generate_library(num_seqs=5, seed=42)

print(df[['seq', 'demo.op.key.positions', 'demo.op.key.marker_tags']])
```
