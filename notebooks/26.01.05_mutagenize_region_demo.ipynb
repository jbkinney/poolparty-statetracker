```python
# Demo: mutagenize() with region parameter
import poolparty as pp
pp.init()

# Test 1: Region by explicit interval [start, stop]
bg1 = pp.from_seq('GGGGAAAAAACCCC')
mut1 = pp.mutagenize(bg1, num_mutations=2, region=[4, 10]).print_library()

# Test 2: Region by marker name  
bg2 = pp.from_seq('GGGG<orf>AAAAAA</orf>CCCC')
mut2 = pp.mutagenize(bg2, num_mutations=2, region='orf').print_library()

# Test 3: Sequential mode with region
bg3 = pp.from_seq('GGGGAAACCCC')
mut3 = pp.mutagenize(bg3, num_mutations=1, region=[4, 7], mode='sequential').print_library()
```
