```python
import poolparty as pp
```

```python
# Pool.mutagenize_region() applies mutagenize() to a marked region (marker removed by default)
with pp.Party():
    bg = pp.from_seq('FLANK<target>ACGT</target>FLANK')
    result = bg.mutagenize_region('target', num_mutations=1, mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# With remove_marker=False, the marker tags are preserved around the mutagenized content
with pp.Party():
    bg = pp.from_seq('FLANK<target>ACGT</target>FLANK')
    result = bg.mutagenize_region('target', remove_marker=False, num_mutations=1, mark_changes=True, mode='sequential')
    df = result.generate_library(num_cycles=1)
df
```

```python
# Works with mutation_rate as well
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGTACGTACGT</region>TTT')
    result = bg.mutagenize_region('region', mutation_rate=0.25)
    df = result.generate_library(num_cycles=5, seed=42)
df
```

```python
# apply_at_marker() also supports remove_marker=False for any transformation
with pp.Party():
    bg = pp.from_seq('AAA<region>ACGT</region>TTT')
    result = bg.apply_at_marker('region', pp.reverse_complement, remove_marker=False)
    df = result.generate_library(num_cycles=1)
df
```
