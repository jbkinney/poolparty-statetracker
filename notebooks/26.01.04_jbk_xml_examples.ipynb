```python
import poolparty as pp
pp.init()

marked_seq = 'TCCCGACT<mut>GGAAAGCGGGCAG<misc/>TGAGCACACAGGAAA</mut>ATTACGG<bc/>AGATCGGA'
pool = pp.from_seq(marked_seq, name='marked_seq')
pool.print_library()

pool = pool.ops.mutagenize('mut',
                           mutation_rate=0.25,
                           mode='hybrid', 
                           num_hybrid_states=3)\
         .ops.repeat_states(5)\
         .ops.get_kmers('bc',
                        length=5,
                        case='lower')\
         .named('processed_seqs')
pool.print_library()

pool = pool.ops.clear_markers().ops.upper().named('final_seqs')
pool.print_library()
```
