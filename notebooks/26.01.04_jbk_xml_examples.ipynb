```python
import poolparty as pp
pp.reset()


marked_seq = 'TCCCGACT.<mut>GGAAAGCGGGCAG<misc/>TGAGCACACAGGAAA</mut>.ATTACGG.<bc/>.AGATCGGA'
pool = pp.from_seq(marked_seq)
print('\nmarked_seq:')
pool.preview_library(num_cycles=1)

pool = pool.ops.mutagenize('mut',
                           mutation_rate=0.25,
                           remove_marker=False,
                           mark_changes=True, 
                           mode='hybrid', 
                           num_hybrid_states=5,
                           swapcase=True)\
         .ops.get_kmers('bc',
                        length=5,
                        remove_marker=False,
                        case='lower')
print('\nmutagenized:')
pool.preview_library(num_cycles=1)

pool = pool.ops.clear_markers()
print('\ncleared markers:')
pool.preview_library(num_cycles=1)
```
