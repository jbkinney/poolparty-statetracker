```python
import poolparty as pp
pp.init()

template = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGCAGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')

wt_pool = (
    template.deepcopy()
    .named('wt_pool')
    .name_seqs(prefix='wt.v')
    .print_library()    
)

mutagenized_pool = (
    template.deepcopy()
    .mutagenize('cre', mutation_rate=0.25, mode='hybrid', num_hybrid_states=5)
    .named('mutagenized_pool')
    .name_seqs(prefix='mut_')
    .repeat_states(3)
    .name_seqs(prefix='v')
    .print_library()
)


final_pool = (
    pp.stack([
        wt_pool,
        mutagenized_pool,
        ])
    .named('barcoded_pool')
    .insert_kmers('bc', length=5, case='lower')
    .clear_nonmolecular_chars()
    .print_library()
)
```
