```python
# Scan functions demo: insertion_scan, replacement_scan, deletion_scan, shuffle_scan
import poolparty as pp
```

```python
# insertion_scan: Insert a sequence at all positions in the background
with pp.Party():
    result = pp.insertion_scan('AAAAAAAAAA', ins_pool='TTT', mode='sequential').named('result')
df = result.generate_library(num_cycles=1)
print(f"insertion_scan: {len(df)} sequences (bg=10, ins=3 -> output length={len(df['seq'].iloc[0])})")
df.head()
```

```python
# replacement_scan: Replace a segment with insert at all positions (preserves length)
with pp.Party():
    result = pp.replacement_scan('AAAAAAAAAA', ins_pool='TTT', mode='sequential').named('result')
df = result.generate_library(num_cycles=1)
print(f"replacement_scan: {len(df)} sequences (output length={len(df['seq'].iloc[0])} = background length)")
df.head()
```

```python
# deletion_scan: Delete a segment at all positions (with gap marker)
with pp.Party():
    result = pp.deletion_scan('AAAAAAAAAA', deletion_length=3, deletion_marker='-', mode='sequential').named('result')
df = result.generate_library(num_cycles=1)
print(f"deletion_scan (with marker): {len(df)} sequences")
df.head()
```

```python
# deletion_scan without marker: segment is removed entirely
with pp.Party():
    result = pp.deletion_scan('AAAAAAAAAA', deletion_length=3, deletion_marker=None, mode='sequential').named('result')
df = result.generate_library(num_cycles=1)
print(f"deletion_scan (no marker): output length={len(df['seq'].iloc[0])} (reduced by 3)")
df.head()
```

```python
# shuffle_scan: Shuffle characters within a window at all positions
with pp.Party():
    result = pp.shuffle_scan('AAACGTAAAA', shuffle_length=3, positions=[3], mode='sequential').named('result')
df = result.generate_library(num_seqs=10, seed=42)
print(f"shuffle_scan: shuffling 'CGT' region -> various permutations")
df
```

```python
# mark_changes: Highlight modified regions with case swap
with pp.Party():
    result = pp.replacement_scan('AAAAAAAAAA', ins_pool='TTT', mark_changes=True, mode='sequential').named('result')
df = result.generate_library(num_cycles=1)
print("replacement_scan with mark_changes=True: replaced region shown in lowercase")
df.head()
```

```python
# spacer_str: Add spacers around the modified region
with pp.Party():
    result = pp.insertion_scan('AAAAAAAAAA', ins_pool='TTT', spacer_str='.', mode='sequential').named('result')
df = result.generate_library(num_cycles=1)
print("insertion_scan with spacer_str='.' adds dots around insert")
df.head()
```

```python
# Random mode: positions selected randomly (default behavior)
with pp.Party():
    result = pp.replacement_scan('AAAAAAAAAA', ins_pool='TTT', mode='random').named('result')
df = result.generate_library(num_seqs=10, seed=42)
print("replacement_scan with mode='random': positions vary randomly")
df
```
