```python
import poolparty as pp; pp.init()

# Demo: Fixed ops with region parameter
pool = pp.from_seq('ACGT<region>ttaa</region>GGCC')
print("Original:"); pool.print_library(show_header=False, show_state=False)

# 1. Case transformations (entire sequence)
print("lower():"); pool.lower().print_library(show_header=False, show_state=False)
print("upper():"); pool.upper().print_library(show_header=False, show_state=False)
print("swapcase():"); pool.swapcase().print_library(show_header=False, show_state=False)

# 2. Case transformations (marker region only, preserve marker)
print("lower(region, keep marker):"); pool.lower('region', remove_marker=False).print_library(show_header=False, show_state=False)
print("upper(region, keep marker):"); pool.upper('region', remove_marker=False).print_library(show_header=False, show_state=False)
print("swapcase(region, keep marker):"); pool.swapcase('region', remove_marker=False).print_library(show_header=False, show_state=False)

# 3. Case transformations (marker region, remove marker)
print("lower(region, remove marker):"); pool.lower('region', remove_marker=True).print_library(show_header=False, show_state=False)

# 4. Reverse complement with region
print("reverse_complement (full):"); pool.reverse_complement().print_library(show_header=False, show_state=False)
print("reverse_complement (region, keep marker):"); pool.reverse_complement('region', remove_marker=False).print_library(show_header=False, show_state=False)
print("reverse_complement (region, remove marker):"); pool.reverse_complement('region', remove_marker=True).print_library(show_header=False, show_state=False)

# 5. Clear gap chars (preserves markers)
pool2 = pp.from_seq('AC-GT<target>TT--AA</target>GG CC')
print("clear_gap_chars - Original:"); pool2.print_library(show_header=False, show_state=False)
print("clear_gap_chars (full):"); pool2.clear_gap_chars().print_library(show_header=False, show_state=False)
print("clear_gap_chars (region only):"); pool2.clear_gap_chars('target', remove_marker=False).print_library(show_header=False, show_state=False)
```
