```python
import poolparty as pp
```

```python
# upper() - convert to uppercase (preserves markers)
with pp.Party():
    pool = pp.from_seq('acgt<region strand="-">tt</region>')
    result = pp.upper(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# lower() - convert to lowercase (preserves markers)
with pp.Party():
    pool = pp.from_seq('ACGT<region strand="-">TT</region>')
    result = pp.lower(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# swapcase() - swap uppercase/lowercase (preserves markers)
with pp.Party():
    pool = pp.from_seq('ACgt<m>XX</m>')
    result = pp.swapcase(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# reverse_complement() - repositions markers based on reversed coordinates
with pp.Party():
    pool = pp.from_seq('ACG<region>TT</region>AA')
    result = pp.reverse_complement(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# clear_ignore_chars() - remove gaps/spaces, keep markers
with pp.Party():
    pool = pp.from_seq('A-C-G-T<m>XX</m>')
    result = pp.clear_ignore_chars(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# clear_ignore_chars() - remove gaps/spaces, keep markers
with pp.Party():
    pool = pp.from_seq('A-C-G-T<m>XX</m>')
    result = pp.clear_markers(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# clear_gap_chars() - remove gaps/spaces, keep markers and alphabet chars
with pp.Party():
    pool = pp.from_seq('A-C-G-T<m>XX</m>')
    result = pp.clear_gap_chars(pool)
result.generate_library(num_cycles=1, seqs_only=True)
```

```python
# OpsContainer convenience methods (pool.ops.*)
with pp.Party():
    pool = pp.from_seq('ACgt')
    rc = pool.ops.reverse_complement()
    sc = pool.ops.swapcase()
    up = pool.ops.upper()
    lo = pool.ops.lower()
print('reverse_complement:', rc.generate_library(num_cycles=1, seqs_only=True))
print('swapcase:', sc.generate_library(num_cycles=1, seqs_only=True))
print('upper:', up.generate_library(num_cycles=1, seqs_only=True))
print('lower:', lo.generate_library(num_cycles=1, seqs_only=True))
```

```python
# clear_gap_chars via OpsContainer
with pp.Party():
    pool = pp.from_seq('A-C-G-T<m>X</m>')
    cg = pool.ops.clear_gap_chars()
print('clear_gap_chars:', cg.generate_library(num_cycles=1, seqs_only=True))
```
