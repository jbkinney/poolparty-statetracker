```python
import poolparty as pp; pp.init()

# Demo: clear_annotation() removes all markers and non-molecular chars, then uppercases
pool = pp.from_seq('ac-gt<region>TT<nested>aa</nested>--CC</region>GG..CC')
print("Original:"); pool.print_library(show_header=False, show_state=False)

# Full sequence clearing (removes ALL markers and non-molecular chars)
print("clear_annotation():"); pool.clear_annotation().print_library(show_header=False, show_state=False)

# Region-specific clearing (clears only inside 'region', keeps outer structure)
print("clear_annotation('region', remove_marker=False):"); pool.clear_annotation('region', remove_marker=False).print_library(show_header=False, show_state=False)
print("clear_annotation('region', remove_marker=True):"); pool.clear_annotation('region', remove_marker=True).print_library(show_header=False, show_state=False)
```
