```python
import poolparty as pp; pp.init()

# Demo 1: Simple operation naming with prefix
print('Demo 1: Simple operation with prefix')
pp.from_seq('ACGTACGT').mutagenize(num_mutations=1, mode='sequential', prefix='mut').print_library(num_cycles=1)

# Demo 2: Chained operations - names ordered by iter_order
print('\nDemo 2: Chained operations')
pp.from_seq('ACGT<bc/>TGCA').mutagenize(
    num_mutations=1, mode='random', num_states=3, prefix='mut'
).insert_kmers(
    region='bc', length=3, mode='sequential', prefix='bc'
).print_library(num_cycles=1, seed=42)

# Demo 3: Stack operation - only active branch contributes
print('\nDemo 3: Stack operation (branch selection)')
pool1 = pp.from_seq('AAAA<bc/>TTTT').mutagenize(num_mutations=1, mode='sequential', prefix='mut1')
pool2 = pp.from_seq('CCCC<bc/>GGGG').mutagenize(num_mutations=1, mode='sequential', prefix='mut2')
pp.stack([pool1, pool2]).insert_kmers(
    region='bc', length=2, mode='sequential', prefix='bc'
).print_library(num_cycles=1)

# Demo 4: Explicit seq_names override
print('\nDemo 4: Explicit seq_names')
pp.from_seqs(['AAA', 'TTT', 'GGG'], seq_names=['wt', 'mut1', 'mut2'], mode='sequential').print_library(num_cycles=1)

# Demo 5: insertion_scan with multiple prefixes
print('\nDemo 5: insertion_scan with multiple prefixes')
sites = pp.from_seqs(['AAA', 'TTT'], mode='sequential', prefix='site')
pp.from_seq('GGG<reg>CCCCCC</reg>GGG').insertion_scan(
    region='reg',
    ins_pool=sites,
    positions=[0, 3],
    mode='sequential',
    prefix='ins',
    prefix_position='pos',
    prefix_insert='site'
).print_library(num_cycles=1)
```
