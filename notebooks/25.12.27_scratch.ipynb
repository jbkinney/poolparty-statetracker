```python
from poolparty.counter import Counter, CounterManager, synchronize_counters

with CounterManager() as mgr:
    A = Counter(5).named('A')
    B = Counter(3).named('B')
    ApB = (A+B).named('ApB')
    
    C = Counter(2).named('C')
    D = Counter(4).named('D')
    CxD = (C*D).named('CxD')
    
    E = synchronize_counters(ApB,CxD).named('E')

mgr.print_graph() 
mgr.test_iteration(E);
```

```python
import poolparty as pp
L=5
with pp.Party() as party:
    A = pp.from_seqs(['A'*L,'C'*L, 'G'*L, 'T'*L], mode='sequential').named('A')
    B = pp.get_kmers(length=L, mode='hybrid', hybrid_mode_num_states=8).named('B')
    A2 = (2*A).named('A2')
    pp.sync([A2,B])
    C = pp.join([A2,B], spacer_str=' ').named('C')

party.print_graph(style='clean')
C.generate_seqs(num_complete_iterations=1, report_op_keys=False)
```
