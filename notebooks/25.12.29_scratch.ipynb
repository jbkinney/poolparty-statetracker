```python
import poolparty as pp
bg_L = 20
ins_L = 5
with pp.Party() as party:
    bg = pp.from_seqs(['A'*bg_L, 'T'*bg_L], 
                      iter_order=2, mode='sequential').named('bg')
    ins = pp.get_kmers(length=ins_L, 
                       mode='hybrid', 
                       num_hybrid_states=3,
                       iter_order=0).named('ins')   
    result = pp.replacement_scan(bg_pool=bg, 
                                 ins_pool=ins, 
                                 spacer_str='.', 
                                 mode='hybrid', 
                                 num_hybrid_states=4, 
                                 op_iter_order=1).named('result')  

pp.generate_seqs(pool=result, num_complete_iterations=1, seed=3)['seq']
```
