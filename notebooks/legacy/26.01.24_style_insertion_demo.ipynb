```python
import poolparty as pp; pp.init()

# style highlights all inserted positions with inline styles
# Similar to style in mutagenize()

# Basic insertion_scan with style
pp.insertion_scan('AAAAAAAAAA', 'GGG', positions=[5], mode='sequential',
                  style='red').print_library()

# replacement_scan also supports style
pp.replacement_scan('AAAAAAAAAA', 'GGG', positions=[5], mode='sequential',
                    style='cyan').print_library()

# Multiple insertion positions - each insert gets styled
pp.insertion_scan('AAAAAAAAAA', 'TT', positions=[0, 5, 10], mode='sequential',
                  style='green bold').print_library()

# Combining style with spacer_str (spacers not styled)
pp.insertion_scan('AAAAAAAAAA', 'GG', positions=[5], mode='sequential',
                  spacer_str='.', style='magenta').print_library()

# Combining with styles on insert pool - both styles apply
pp.insertion_scan('AAAAAAAAAA', 
                  pp.from_seq('CCC').stylize(style='underline'),
                  positions=[5], mode='sequential',
                  style='yellow').print_library()
```
