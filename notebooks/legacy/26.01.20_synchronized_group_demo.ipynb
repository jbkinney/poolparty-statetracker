```python
# SynchronizedGroup demo - states that share the same value
from statetracker import Manager, State, product, sync, synced_to

with Manager() as m:
    # Test 1: Basic synced states - bidirectional value setting
    print("=== Test 1: Bidirectional sync ===")
    A = State(5, name='A')
    B = synced_to(A, name='B')
    print(f"Initial: A={A.value}, B={B.value}")
    A.value = 3
    print(f"After A.value=3: A={A.value}, B={B.value}")
    B.value = 1  # Setting B also updates A!
    print(f"After B.value=1: A={A.value}, B={B.value}")
    
    # Test 2: Synced states with product - B gets updated when D is set
    print("\n=== Test 2: D=product([A,C]) with A-B synced ===")
    A2 = State(3, name='A2')
    B2 = synced_to(A2, name='B2')
    C2 = State(4, name='C2')
    D2 = product([A2, C2]).named('D2')
    D2.value = 5
    print(f"After D2.value=5: A2={A2.value}, B2={B2.value}, C2={C2.value}, D2={D2.value}")
    print(f"A2 and B2 synchronized: {A2.value == B2.value}")
    
    # Test 3: Transitive merging - sync(A,B) + sync(B,C) => A,B,C all synced
    print("\n=== Test 3: Transitive merging ===")
    X = State(5, name='X')
    Y = State(5, name='Y')
    Z = State(5, name='Z')
    sync(X, Y)
    sync(Y, Z)
    X.value = 4
    print(f"After X.value=4: X={X.value}, Y={Y.value}, Z={Z.value}")
    
    # Test 4: print_dag shows synced peers directly
    print("\n=== Test 4: print_dag output ===")
    print("A.print_dag():")
    A.print_dag()
    print("\nB.print_dag():")
    B.print_dag()
```
