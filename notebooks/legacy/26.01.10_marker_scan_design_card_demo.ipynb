```python
import poolparty as pp; pp.init()

# Demo: MarkerScanOp design card includes position_index, start, stop, length, region_name, region_content, strand, region
seq = "ACGTACGTACGT"

# Zero-length marker scan (start=stop, length=0, region_content='')
pool = pp.marker_scan(seq, marker='site', marker_length=0, mode='sequential')
df = pool.generate_library(report_design_cards=True)
display(df)
```

```python
# Region marker scan (start < stop, length=3, region_content shows the sequence)
pool2 = pp.marker_scan(seq, marker='region', marker_length=3, mode='sequential')
df = pool2.generate_library(report_design_cards=True)
display(df)
```
