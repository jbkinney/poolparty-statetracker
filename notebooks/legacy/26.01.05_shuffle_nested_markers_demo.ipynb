```python
# Demonstrates shuffle_seq with nested markers - only molecular characters are shuffled
import poolparty as pp
pp.init()

# Template with nested markers: <cre> contains <misc/> inside
template_pool = pp.from_seq('TCCCGACT<cre>GCTCTCTCTCTA<misc/>AGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')

result_pool = (
    template_pool
    .shuffle_seq('cre', 
                 remove_marker=True, 
                 mode='hybrid', 
                 mark_changes=True,
                 num_hybrid_states=5,
                 name_prefix='shuffle_')
    .repeat_states(3, name_prefix='v')
    .print_tree()
    .print_library()
)
# Note: <misc/> marker preserved at original position, only DNA bases shuffled
```
