```python
import poolparty as pp

# Demo: deletion_scan - scans deletions at each position
bg = 'AAACCCCTTTT'

# Basic deletion: replace 3 chars with '---' at each position
with pp.Party():
    print('=== deletion_scan (default deletion_marker="-") ===')
    del_scan = pp.deletion_scan(bg, deletion_length=3, mode='sequential')
    del_scan.print_library()

# Without marker: segment is simply removed (empty string)
with pp.Party():
    print('\n=== deletion_scan with deletion_marker=None ===')
    pp.deletion_scan(bg, deletion_length=3, deletion_marker=None, mode='sequential').print_library()

# With custom marker character
with pp.Party():
    print('\n=== deletion_scan with custom deletion_marker="X" ===')
    pp.deletion_scan(bg, deletion_length=2, deletion_marker='X', mode='sequential').print_library()

# Deletion within a region - marker_scan handles region internally
with pp.Party():
    bg_with_region = 'AAAA<region>CCCCCCCC</region>TTTT'
    print('\n=== deletion_scan within region ===')
    pp.deletion_scan(bg_with_region, deletion_length=3, region='region', mode='sequential').print_library()

# Underlying mechanism: marker_scan + from_seq with region
with pp.Party():
    print('\n=== Manual approach: marker_scan + from_seq ===')
    marked = pp.marker_scan('AAACCCCTTTT', marker='_del', marker_length=3, mode='sequential')
    pp.from_seq('---', bg_pool=marked, region='_del', remove_marker=True).print_library()
```
