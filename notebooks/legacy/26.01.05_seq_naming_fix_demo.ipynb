```python
import poolparty as pp
```

```python
# Demonstrate that replacement_scan now correctly combines parent names
pp.init()
template = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGCAGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')

# Create motif pool with explicit sequence names
motif_pool = pp.from_seqs(
    ['A'*5, 'T'*5], 
    seq_names=['motif_a', 'motif_t'],
    mode='sequential',
)

# Use replacement_scan with name_prefix
replacement_pool = (
    template
    .replacement_scan('cre',
                      ins_pool=motif_pool,
                      positions=slice(None, None, 5),
                      mode='sequential', 
                      name_prefix='rep_')
)

# The resulting names should combine 'rep_X' with 'motif_a' or 'motif_t'
replacement_pool.insert_kmers('bc', length=5, case='lower').print_library()
```

```python
# Demonstrate insertion_scan also combines names correctly
pp.init()
bg = pp.from_seq('ACGT' * 5)

insert_pool = pp.from_seqs(
    ['XXX', 'YYY'],
    seq_names=['insert_x', 'insert_y'],
    mode='sequential',
)

result = pp.insertion_scan(
    bg, 
    insert_pool, 
    positions=[0, 10], 
    mode='sequential',
    name_prefix='pos_',
)

result.print_library()
```

```python
# Demonstrate join also combines all parent names
pp.init()
pool_a = pp.from_seqs(['AAA', 'BBB'], seq_names=['a1', 'a2'], mode='sequential')
pool_b = pp.from_seqs(['CCC', 'DDD'], seq_names=['b1', 'b2'], mode='sequential')
pool_c = pp.from_seqs(['EEE', 'FFF'], seq_names=['c1', 'c2'], mode='sequential')

joined = pp.join([pool_a, pool_b, pool_c])
joined.print_library()
```
