```python
import poolparty as pp; pp.init()

# Demo: report_design_cards parameter controls output format
pool = pp.from_seqs(['AAA', 'TTT', 'CCC'], seq_names=['seq_a', 'seq_b', 'seq_c']).named('myseq')
mutants = pp.mutagenize(pool, num_mutations=1).named('mutants')

# Default (report_design_cards=False): minimal output with just "name" and "seq" columns
print("Default (report_design_cards=False):")
df_minimal = mutants.generate_library(num_seqs=3)
print(df_minimal)
print()

# With report_design_cards=True: full design card information
print("With report_design_cards=True:")
df_full = mutants.generate_library(num_seqs=3, report_design_cards=False)
print(df_full)
```
