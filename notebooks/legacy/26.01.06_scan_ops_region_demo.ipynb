```python
import poolparty as pp; pp.init()

# Demo: scan_ops with region support
# Each scan operation now supports region, remove_marker, and mark_changes parameters

bg = 'AAAA<region>CCCCCCCC</region>TTTT'  # 16 total molecular chars, 8 in region

# 1. insertion_scan within a region - insert 'GG' at positions within the marked region
print('=== insertion_scan with region ===')
pp.insertion_scan(bg, 'GG', region='region', mark_changes=True, mode='sequential').print_library(num_cycles=1)

# 2. deletion_scan within a region - delete 2 chars at positions within the marked region
print('\n=== deletion_scan with region ===')
pp.deletion_scan(bg, deletion_length=2, region='region', deletion_marker='-', mode='sequential').print_library(num_cycles=1)

# 3. replacement_scan within a region - replace 2 chars with 'GG' at positions within the marked region
print('\n=== replacement_scan with region ===')
pp.replacement_scan(bg, 'GG', region='region', mark_changes=True, mode='sequential').print_library(num_cycles=1)

# 4. mutagenize_scan within a region - mutagenize 3 chars at positions within the marked region
print('\n=== mutagenize_scan with region ===')
pp.mutagenize_scan(bg, mutagenize_length=3, num_mutations=1, region='region', mark_changes=True, mode='sequential').print_library(num_seqs=5, seed=42)

# 5. shuffle_scan within a region - shuffle 3 chars at positions within the marked region
print('\n=== shuffle_scan with region ===')
pp.shuffle_scan(bg, shuffle_length=3, region='region', mark_changes=True, mode='sequential').print_library(num_seqs=5, seed=42)

# 6. subseq_scan within a region - extract 3-char subsequences from within the marked region
print('\n=== subseq_scan with region ===')
pp.subseq_scan(bg, seq_length=3, region='region', mode='sequential').print_library(num_cycles=1)

# 7. Using [start, stop] interval instead of marker name
print('\n=== replacement_scan with [start, stop] region ===')
bg_no_marker = 'AAAACCCCCCCCTTTT'
pp.replacement_scan(bg_no_marker, 'GG', region=[4, 12], mark_changes=True, mode='sequential').print_library(num_cycles=1)
```
