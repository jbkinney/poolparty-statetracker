```python
import poolparty as pp; pp.init()

# Demo: Slicing and stacking pools works correctly
# Previously, slicing a pool then stacking would produce incorrect results

# Create a pool with 4 sequences
pool = pp.from_seqs(['A', 'B', 'C', 'D'], mode='sequential')
print("Original pool:")
pool.print_library()

# Slice into first and last halves
first = pool[:2]  # A, B
last = pool[-2:]  # C, D

# Stack them back together
combined = (first + last).named('combined')
print("\nSliced and stacked (first[:2] + last[-2:]):")
combined.print_library()

# Also works with independent pools
a = pp.from_seqs(['X', 'Y'], mode='sequential')
b = pp.from_seqs(['Z', 'W'], mode='sequential')
stacked = (a + b).named('stacked')
print("\nTwo independent pools stacked:")
stacked.print_library()
```
