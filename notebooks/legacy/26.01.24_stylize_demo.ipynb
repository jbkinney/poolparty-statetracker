```python
# Demo of stylize() - apply inline styling without modifying sequences
import poolparty as pp
pp.init()

# Basic stylize - style all contents red
pool = pp.from_seq('AAAA<test>GGGGCCCC</test>TTTT').named('template')
pool.stylize(style='red').print_library()

# Stylize with region - only style the marker region (region is first positional arg)
pool.stylize('test', style='cyan bold').print_library()

# Stylize with which='upper' - only style uppercase chars
pp.from_seq('AcGtAcGt').stylize(style='blue', which='upper').print_library()

# Stylize with which='lower' - only style lowercase chars  
pp.from_seq('AcGtAcGt').stylize(style='green', which='lower').print_library()

# Stylize with interval region [start, stop]
pp.from_seq('AACCCCGG').stylize([2, 6], style='magenta').print_library()

# Chaining multiple stylize operations
pp.from_seq('AcGt').stylize(style='red bold', which='upper').stylize(style='cyan', which='lower').print_library()

# Case transforms: 'lower'/'lowercase', 'upper'/'uppercase', 'swapcase'
pp.from_seq('ACGT').stylize(style='lowercase cyan bold').print_library()
pp.from_seq('acgt').stylize(style='uppercase red').print_library()
pp.from_seq('AcGt').stylize(style='swapcase magenta').print_library()

# Stylize after mutagenize to highlight mutations
pool.mutagenize('test', mutation_rate=0.3, mode='random', num_states=3).stylize(style='yellow underline', which='lower').print_library(seed=42)

# Invalid styles raise a helpful error when styles are rendered
try:
    pp.from_seq('ACGT').stylize(style='foobar').print_library()
except ValueError as e:
    print(f"ValueError: {e}")
```
