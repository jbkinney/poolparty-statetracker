```python
import poolparty as pp; pp.init()

# Demo: deletion_scan - directly scans and deletes sequences (refactored implementation)
with pp.Party():
    bg = pp.from_seq('AAACCCCTTTT')
    
    # Basic deletion: replace 3 chars with '---' at each position
    bg.deletion_scan(deletion_length=3, mode='sequential').print_library()
    
    # Without marker: segment is simply removed
    bg.deletion_scan(deletion_length=3, deletion_marker=None, mode='sequential').print_library()
    
    # With custom marker character
    bg.deletion_scan(deletion_length=2, deletion_marker='X', mode='sequential').print_library()
    
    # Deletion within a region
    bg_with_region = pp.from_seq('AAAA<region>CCCCCCCC</region>TTTT')
    bg_with_region.deletion_scan(deletion_length=3, region='region', mode='sequential').print_library()
    
    # With style parameter
    bg.deletion_scan(deletion_length=3, style='red bold', mode='sequential').print_library()
    
    # Random mode
    bg.deletion_scan(deletion_length=3, mode='random', num_states=5).print_library()
```
