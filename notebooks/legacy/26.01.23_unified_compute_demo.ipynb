```python
import poolparty as pp; pp.init()
import numpy as np

# Demo: Unified compute method returns both design card and sequences together
pool = pp.from_seqs(["AAA", "TTT", "CCC"], seq_names=["seq_a", "seq_b", "seq_c"]).named("myseq")
mutants = pp.mutagenize(pool, num_mutations=1).named("mutants")

# Generate library with design cards - now design card info is always available
df = mutants.generate_library(num_seqs=5, seed=42, report_design_cards=True)
print("Generated sequences with design card information:")
print(df[["name", "seq", "mutants.key.positions", "mutants.key.wt_chars", "mutants.key.mut_chars"]].head())
print()

# Demonstrate direct compute method usage - returns combined dict
print("Direct compute() call returns both design card and sequence:")
result = mutants.operation.compute(["AAA"], rng=np.random.default_rng(42))
print(f"Design card keys: {[k for k in result.keys() if not k.startswith('seq_')]}")
print(f"Sequence keys: {[k for k in result.keys() if k.startswith('seq_')]}")
print(f"Full result: {result}")
```
