```python
import poolparty as pp; pp.init()

# style parameter on source operations - styles all positions

# from_seq with style
pp.from_seq('ACGT', style='red').print_library()

# from_seqs with style
pp.from_seqs(['ACGT', 'TTTT'], mode='sequential', style='blue').print_library()

# get_kmers with style
pp.get_kmers(length=3, mode='sequential', style='green').print_library()

# style_background - styles non-modified positions

# mutagenize with style_mutations and style_background
pp.mutagenize('ACGTACGT', num_mutations=1, mode='sequential',
              style_mutations='red', style_background='cyan').print_library()

# insertion_scan with style_insertion and style_background
pp.insertion_scan('AAAAAAAAAA', 'GGG', positions=[5], mode='sequential',
                  style_insertion='red', style_background='blue').print_library()

# deletion_scan with style_deletion and style_background
pp.deletion_scan('AAAAAAAAAA', deletion_length=2, positions=[5], mode='sequential',
                 style_deletion='gray', style_background='cyan').print_library()

# insert_kmers with style_kmers and style_background
pp.from_seq('AA<kmer/>TT').insert_kmers('kmer', length=2, mode='sequential',
                                         style_kmers='red', style_background='blue').print_library()
```
