```python
# Marker Scan Demo - Insert placeholder tags into sequences
import poolparty as pp
```

```python
# Basic marker insertion - random position
with pp.Party() as party:
    pool = pp.marker_scan('ACGTACGT', marker='insert_site')

pool.generate_library(num_seqs=5, seed=42)
```

```python
# Sequential mode - enumerate all insertion positions
with pp.Party() as party:
    pool = pp.marker_scan('ACGT', marker='m', mode='sequential')

pool.generate_library(num_cycles=1)
```

```python
# Custom positions - limit where markers can be inserted
with pp.Party() as party:
    pool = pp.marker_scan('ACGTACGT', marker='site', positions=[0, 4, 8], mode='sequential')

pool.generate_library(num_cycles=1)
```

```python
# Using Marker objects directly
with pp.Party() as party:
    marker = pp.Marker(name='promoter')
    print(f"Marker: {marker}")
    print(f"Tag: {marker.tag}")
    pool = pp.marker_scan('ACGTACGT', marker=marker)

pool.generate_library(num_seqs=3, seed=42)
```

```python
# Alphabet methods handle markers correctly
alpha = pp.get_alphabet('dna')
seq_with_marker = 'AC{marker}GT'

print(f"Sequence: {seq_with_marker}")
print(f"Actual DNA length: {alpha.get_seq_length(seq_with_marker)}")  # Excludes marker
print(f"Valid positions: {alpha.get_valid_seq_positions(seq_with_marker)}")  # Excludes marker positions
```

```python
# Combine with other operations
with pp.Party() as party:
    marked = pp.marker_scan('ACGTACGT', marker='insert', mode='sequential')
    combined = pp.join(['5_FLANK_', marked, '_3_FLANK'])

combined.generate_library(num_seqs=5)
```

```python
# Mark sequences multiple times.
with pp.Party() as party:
    seqs = pp.from_seqs(['A'*10, 'T'*10])
    marked1 = pp.marker_scan(seqs, marker='ins1')
    marked2 = pp.marker_scan(marked1, marker='ins2')

marked2.print_tree()
marked2.generate_library(num_seqs=20, seed=42, report_pool_seqs=False, report_pool_states=False, report_op_states=False)
```
