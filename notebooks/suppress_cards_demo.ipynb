```python
import poolparty as pp
pp.init()

# Create a pool with mutagenize operation
pool = pp.from_seq('ATCGATCGATCG').mutagenize(num_mutations=2, mode='sequential', prefix='mut')

# Generate library with design cards (default)
print("=== With design cards (default) ===")
df_with_cards = pool.generate_library(num_cycles=1, report_design_cards=True)
print(f"Number of columns: {len(df_with_cards.columns)}")
print(f"Columns with '.key.' (design cards): {[c for c in df_with_cards.columns if '.key.' in c]}")
print(f"\nFirst 3 sequences:")
print(df_with_cards[['seq']].head(3))

# Toggle cards off - suppresses design card computation at the source
pp.toggle_cards(on=False)

# Generate library with cards suppressed - no design card keys
print("\n=== With pp.toggle_cards(on=False) ===")
df_no_cards = pool.generate_library(num_cycles=1, report_design_cards=True)
print(f"Number of columns: {len(df_no_cards.columns)}")
print(f"Columns with '.key.' (design cards): {[c for c in df_no_cards.columns if '.key.' in c]}")
print(f"\nFirst 3 sequences:")
print(df_no_cards[['seq']].head(3))

# Verify styles are still computed (toggle_cards is independent of toggle_styles)
print(f"\nInline styles preserved: {df_no_cards['_inline_styles'].iloc[0] is not None}")

# Toggle cards back on
pp.toggle_cards(on=True)

# Summary
print(f"\n=== Summary ===")
print(f"With design cards: {len(df_with_cards.columns)} columns")
print(f"With toggle_cards(on=False): {len(df_no_cards.columns)} columns")
print(f"Design card keys eliminated: {len([c for c in df_with_cards.columns if '.key.' in c])} columns removed")
```
