```python
import poolparty as pp
pp.init()

# Create a pool with mutagenize operation
pool = pp.from_seq('ATCGATCGATCG').mutagenize(num_mutations=2, mode='sequential', prefix='mut')

# Generate library without suppress_cards (default) - includes design card info
print("=== With design cards (default) ===")
df_with_cards = pool.generate_library(num_cycles=1, report_design_cards=True)
print(f"Number of columns: {len(df_with_cards.columns)}")
print(f"Columns with '.key.' (design cards): {[c for c in df_with_cards.columns if '.key.' in c]}")
print(f"\nFirst 3 sequences with names:")
print(df_with_cards[['name', 'seq']].head(3))

# Generate library with suppress_cards=True - no design card computation
print("\n=== With suppress_cards=True ===")
df_no_cards = pool.generate_library(num_cycles=1, report_design_cards=True, suppress_cards=True)
print(f"Number of columns: {len(df_no_cards.columns)}")
print(f"Columns with '.key.' (design cards): {[c for c in df_no_cards.columns if '.key.' in c]}")
print(f"\nFirst 3 sequences with names:")
print(df_no_cards[['name', 'seq']].head(3))

# Summary
print(f"\n=== Summary ===")
print(f"With design cards: {len(df_with_cards.columns)} columns")
print(f"With suppress_cards: {len(df_no_cards.columns)} columns")
print(f"Design card overhead eliminated: {len(df_with_cards.columns) - len(df_no_cards.columns)} columns removed")

# Print library also supports suppress_cards
print("\n=== Using print_library() with suppress_cards ===")
pool.print_library(num_seqs=3, suppress_cards=True)
```
