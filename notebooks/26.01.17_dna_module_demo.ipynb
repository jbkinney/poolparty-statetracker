```python
import poolparty as pp; pp.init()

# Show DNA constants are accessible
print(f"Bases: {pp.BASES}")
print(f"Complement of A: {pp.COMPLEMENT['a']}")
print(f"IUPAC N expands to: {pp.IUPAC_TO_DNA['N']}")
print(f"Ignore chars include: {'-' in pp.IGNORE_CHARS}, {'.' in pp.IGNORE_CHARS}")

# Show IUPAC works
print("\nIUPAC expansion:")
pp.from_iupac('ACGTNNN').print_library()

# Show gaps are handled
print("\nMutations with gaps:")
pp.from_seq('AC-GT').ops.mutagenize(num_mutations=1).print_library()
```
