```python
import poolparty as pp; pp.init()

# shuffle_scan: shuffle characters within a window at scanning positions
# Note: pp.init() sets mark_changes=True by default, so shuffled regions appear lowercase
print("=== shuffle_scan (shuffle 3-char windows) ===")
pp.shuffle_scan('ACGTACGTAC', shuffle_length=3, positions=[0, 4, 7], mode='sequential').print_library()

pp.init()
# Multiple shuffles per position
print("=== shuffle_scan with shuffles_per_position=3 ===")
pp.shuffle_scan('ACGTACGTAC', shuffle_length=3, positions=[0], shuffles_per_position=3).print_library()

pp.init()
# With region constraint
print("=== shuffle_scan with region ===")
pp.shuffle_scan('AAA<target>ACGTG</target>AA', shuffle_length=2, region='target', mode='sequential').print_library()

pp.init()
# mark_changes=False preserves case
print("=== shuffle_scan with mark_changes=False ===")
pp.shuffle_scan('ACGTACGTAC', shuffle_length=3, positions=[5], mark_changes=False).print_library()

pp.init()
# spacer_str adds spacing around shuffled region
print("=== shuffle_scan with spacer_str ===")
shuffled_pool = pp.shuffle_scan('ACGTACGTAC', shuffle_length=3, positions=[5], spacer_str='.').print_library()
```
