```python
import poolparty as pp; pp.init()

# Demonstrate stateless random pools (mode='random' with num_states=None)
# Pure random pool - no state
random_pool = pp.mutagenize('ACGTACGT', num_mutations=1, mode='random').named('pure_random')
print(f"Pure random pool: pool.state is None = {random_pool.state is None}")
print(f"Pure random pool: pool.num_states = {random_pool.num_states}")
print(f"Pure random pool: op.num_values = {random_pool.operation.num_values}")
random_pool.print_library(num_seqs=5, seed=42)

# Random pool with explicit num_states - has state
random_num_states_pool = pp.mutagenize('ACGTACGT', num_mutations=1, mode='random', num_states=10).named('random_num_states')
print(f"\nRandom with num_states: pool.state is None = {random_num_states_pool.state is None}")
print(f"Random with num_states: pool.num_states = {random_num_states_pool.num_states}")
print(f"Random with num_states: op.num_values = {random_num_states_pool.operation.num_values}")
random_num_states_pool.print_library(num_cycles=1, seed=42)

# Sequential parent + random child (stateless)
seq_parent = pp.from_seqs(['AAAA', 'TTTT'], mode='sequential').named('seq_parent')
random_child = pp.mutagenize(seq_parent, num_mutations=1, mode='random').named('random_child')
print(f"\nSequential parent + random child:")
print(f"Parent num_states = {seq_parent.num_states}")
print(f"Child num_states = {random_child.num_states} (passthrough from parent)")
print(f"Child op.num_values = {random_child.operation.num_values}")
random_child.print_library(num_cycles=1, seed=42)

# Random parent (stateless) + sequential child
random_parent = pp.from_seqs(['AAAA', 'TTTT'], mode='random').named('random_parent')
seq_child = pp.mutagenize(random_parent, num_mutations=1, mode='sequential').named('seq_child')
print(f"\nRandom parent + sequential child:")
print(f"Parent num_states = {random_parent.num_states}")
print(f"Child num_states = {seq_child.num_states} (from child op only)")
print(f"Child op.num_values = {seq_child.operation.num_values}")
seq_child.print_library(num_cycles=1, seed=42)
```
