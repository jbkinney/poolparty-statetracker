```python
import poolparty as pp
```

```python
# Pool.apply_at_marker() allows chaining transformations on marked regions
with pp.Party():
    bg = pp.from_seq('ACGT<orf>ATGCCC</orf>TTTT')
    result = bg.apply_at_marker('orf', pp.reverse_complement)
    df = result.generate_library(num_cycles=1)
df['seq'][0]
```

```python
# Chaining multiple apply_at_marker() calls
with pp.Party():
    bg = pp.from_seq('AAA<region1>ACGT</region1>TTT<region2>GGCC</region2>CCC')
    result = (bg
        .apply_at_marker('region1', pp.reverse_complement)
        .apply_at_marker('region2', pp.reverse_complement))
    df = result.generate_library(num_cycles=1)
df['seq'][0]
```

```python
# Chaining with lambda for more complex transformations
with pp.Party():
    bg = pp.from_seq('FLANK<target>ACGT</target>FLANK')
    result = bg.apply_at_marker('target', lambda p: pp.mutagenize(p, num_mutations=1, mark_changes=True))
    df = result.generate_library(num_cycles=1)
df['seq'][0]
```
