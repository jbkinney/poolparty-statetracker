```python
# Basic usage: reverse complement a DNA sequence
import poolparty as pp

seq = 'ATGCAATTGG'

with pp.Party() as party:
    rc_pool = pp.reverse_complement(seq).named('rc')

df = rc_pool.generate_library(num_seqs=1)
print(f"Original:           5'-{seq}-3'")
print(f"Reverse complement: 5'-{df['seq'].iloc[0]}-3'")
```

```python
# Chaining with other operations: mutagenize then reverse complement
with pp.Party() as party:
    mutants = pp.mutagenize('ATGCATGC', mutation_rate=0.3, mode='hybrid', num_hybrid_states=10).named('mut')
    rc_mutants = pp.reverse_complement(mutants).named('rc_mut')

print(f"Original sequence: ATGCATGC")
print(f"Number of 1-mutation variants: {mutants.num_states}")
df = rc_mutants.generate_library(num_cycles=1)
df[['rc_mut.seq', 'mut.seq', 'mut.op.key.positions', 'mut.op.key.wt_chars', 'mut.op.key.mut_chars']]
```

```python
# Works with lowercase and ignore chars (gaps)
with pp.Party() as party:
    rc_lower = pp.reverse_complement('atgc').named('rc')
    df = rc_lower.generate_library(num_seqs=1)
print(f"Lowercase: atgc -> {df['seq'].iloc[0]}")

with pp.Party() as party:
    rc_gapped = pp.reverse_complement('ATG-CAT').named('rc')
    df = rc_gapped.generate_library(num_seqs=1)
print(f"With gap:  ATG-CAT -> {df['seq'].iloc[0]}")
```
