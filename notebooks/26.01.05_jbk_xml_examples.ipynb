```python
import poolparty as pp
pp.init()

template = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGCAGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')

# Use seq_name_prefix directly in repeat() instead of name_seqs()
wt_pool = (
    template
    .repeat_states(5, seq_name_prefix='wt.v')  
)

motif_pool = pp.from_seqs(
    ['A'*5, 'T'*5], 
    seq_names=['motif_a', 'motif_t'],
    mode='sequential',
    )   

replacement_pool = (
    template
    .replacement_scan('cre',
                      ins_pool=motif_pool,
                      positions=slice(None,None,5),
                      mode='sequential', seq_name_prefix='rep_')
)

final_pool = (
    pp.stack([
        wt_pool,
        # mutagenized_pool,
        # deletion_pool,
        replacement_pool,
        ])
    .named('barcoded_pool')
    .insert_kmers('bc', length=5, case='lower')
    .print_library()
)
```
