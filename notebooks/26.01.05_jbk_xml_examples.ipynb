```python
import poolparty as pp
pp.init()

template_pool = pp.from_seq('TCCCGACT<cre>GGAAAGCGGGC<misc/>AGTGAGCACACAGGAAA</cre>ATTACGG<bc/>AGATCGGA')

result_pool = (
    template_pool
    .shuffle_seq('cre', 
                 remove_marker=False, 
                 mode='hybrid', 
                 mark_changes=True,
                 num_hybrid_states=5,
                 seq_name_prefix='shuffle_')
    .repeat_states(3, seq_name_prefix='v')
    .print_tree()
    .print_library(seed=42)
)
```
