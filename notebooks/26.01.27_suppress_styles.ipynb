```python
import poolparty as pp
pp.init()

# Demo: suppress_styles parameter to disable style computation and rendering

# Create a pool with styles
pool = (pp.from_seq('ACGTACGTACGT', style='red')
        .mutagenize(num_mutations=2, style='blue')
        .shuffle_seq(style='green'))

print("With styles (default):")
pool.print_library(num_seqs=3, seed=42)

print("\nWithout styles (suppress_styles=True):")
pool.print_library(num_seqs=3, seed=42, suppress_styles=True)

# Also works with generate_library()
df_with_styles = pool.generate_library(num_seqs=1, seed=42, report_design_cards=True)
df_without_styles = pool.generate_library(num_seqs=1, seed=42, report_design_cards=True, suppress_styles=True)

print(f"\nWith styles: {len(df_with_styles.iloc[0]['_inline_styles'].style_list)} style layers")
print(f"Without styles: {len(df_without_styles.iloc[0]['_inline_styles'].style_list)} style layers")
```
