```python
import poolparty as pp; pp.init()
pp.set_default('mark_changes',True)

# Background sequence and insert
bg = 'AAAAAAAAAA'  # 10 chars
ins = 'TTT'        # 3 chars

# insertion_scan: INSERT at positions (output length = bg + ins = 13)
# Note: pp.init() sets mark_changes=True by default, so inserts appear lowercase
print("=== insertion_scan (insert mode, default) ===")
ins_pool = pp.insertion_scan(bg, ins, positions=[0, 5, 10], mode='sequential', spacer_str='.')
ins_pool.print_library()

# insertion_scan with replace=True: REPLACE at positions (output length = bg = 10)
print("\n=== insertion_scan(replace=True) ===")
rep_pool = pp.insertion_scan(bg, ins, positions=[0, 4, 7], replace=True, mode='sequential')
rep_pool.print_library()

# replacement_scan is now a thin wrapper for insertion_scan(replace=True)
print("\n=== replacement_scan (equivalent to replace=True) ===")
pp.replacement_scan(bg, ins, positions=[0, 4, 7], mode='sequential').print_library()

# mark_changes=False to preserve original case
print("\n=== insertion_scan with mark_changes=False ===")
pp.insertion_scan(bg, ins, positions=[5], mark_changes=False).print_library()

# spacer_str adds spacing around insert
print("\n=== insertion_scan with spacer_str ===")
pp.insertion_scan(bg, ins, positions=[5], spacer_str='.').print_library()
```
