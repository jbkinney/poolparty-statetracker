```python
import poolparty as pp; pp.init()

print("=== DnaPool: DNA-specific operations ===")
dna = pp.from_seq("ATGAAATGA")
print(f"Type: {type(dna).__name__}")
dna.print_library()

print("DnaPool.rc() - reverse complement")
dna.rc().print_library()

print("DnaPool.translate() - returns ProteinPool")
protein = dna.translate()
print(f"Type: {type(protein).__name__}")
protein.print_library()

print("=== ProteinPool: 3-letter amino acid output ===")
protein.print_library(chars_per_aa=3, aa_separator="-")

print("=== DnaPool.mutagenize() returns DnaPool ===")
mutated_dna = pp.from_seq("ACGT").mutagenize(num_mutations=1, num_states=4, mode="sequential")
print(f"Type: {type(mutated_dna).__name__}")
mutated_dna.print_library()

print("=== State operations preserve pool type ===")
print("ProteinPool * 2 returns ProteinPool")
repeated = protein * 2
print(f"Type: {type(repeated).__name__}")
repeated.print_library()

print("=== Pool hierarchy ===")
print(f"Pool base class: {pp.Pool.__name__}")
print(f"DnaPool inherits from Pool: {issubclass(pp.DnaPool, pp.Pool)}")
print(f"ProteinPool inherits from Pool: {issubclass(pp.ProteinPool, pp.Pool)}")
```
