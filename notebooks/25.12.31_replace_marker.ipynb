```python
import poolparty as pp
```

```python
# Basic replace_marker usage: replace a marker in a background sequence
with pp.Party():
    bg = pp.from_seq('AAAA{slot}TTTT')
    ins = pp.from_seq('GGCC')
    result = pp.replace_marker(bg, ins, 'slot')  # string creates Marker
    df = result.generate_seqs(num_seqs=1)
df
```

```python
# Counter composition: ordered_product of bg and ins counters
with pp.Party():
    m = pp.Marker(name='ins')
    bg = pp.from_seqs(['X{ins}', 'Y{ins}'], mode='sequential')
    ins = pp.from_seqs(['AA', 'BB', 'CC'], mode='sequential')
    result = pp.replace_marker(bg, ins, m)
    print(f"bg states: {bg.num_states}, ins states: {ins.num_states}")
    print(f"result states: {result.num_states} (product of parent states)")
    df = result.generate_seqs(num_cycles=1)
df
```

```python
# Combining with marker_scan: insert marker at all positions, then replace
with pp.Party():
    m = pp.Marker(name='site')
    scanned = pp.marker_scan('ACGT', marker=m, mode='sequential')
    insert = pp.from_seqs(['XX', 'YY'], mode='sequential')
    result = pp.replace_marker(scanned, insert, m)
    print(f"scan positions: {scanned.num_states}, inserts: {insert.num_states}")
    print(f"result states: {result.num_states}")
    df = result.generate_seqs(num_cycles=1)
df
```

```python
# All string inputs: pools auto-created, marker created from string name
with pp.Party():
    result = pp.replace_marker('AAA{X}TTT', 'GGG', 'X')
    df = result.generate_seqs(num_seqs=1)
df
```
