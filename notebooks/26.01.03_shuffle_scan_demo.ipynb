```python
import poolparty as pp
```

```python
# Basic shuffle_scan: shuffle a 4-bp window across the background sequence
with pp.Party():
    result = pp.shuffle_scan('AAACGTCGTAAA', shuffle_length=4, mode='sequential').named('result')

df = result.generate_seqs(num_seqs=10, seed=42)
df
```

```python
# shuffle_scan with specific positions
with pp.Party():
    result = pp.shuffle_scan(
        'AAACGTCGTAAA',
        shuffle_length=4,
        positions=[0, 4, 8],
        mode='random'
    ).named('result')

df = result.generate_seqs(num_seqs=10, seed=42)
df
```

```python
# shuffle_scan with mark_changes to highlight the shuffled region
with pp.Party():
    result = pp.shuffle_scan(
        'AAACGTCGTAAA',
        shuffle_length=4,
        mark_changes=True
    ).named('result')

df = result.generate_seqs(num_seqs=10, seed=42)
df
```

```python
# Using scan() directly with action='shuffle'
with pp.Party():
    result = pp.scan(
        action='shuffle',
        bg_pool='AAACGTCGTAAA',
        shuffle_length=6,
        positions=[3],
        mark_changes=True
    ).named('result')

df = result.generate_seqs(num_seqs=10, seed=42)
df
```

```python
# Verify shuffled region contains same characters (just reordered)
with pp.Party():
    result = pp.shuffle_scan(
        'AAACGTCGTAAA',
        shuffle_length=4,
        positions=[3]
    ).named('result')

df = result.generate_seqs(num_seqs=20, seed=42)
shuffled_regions = [seq[3:7] for seq in df['seq']]
print("Shuffled regions (all should contain same chars as 'CGTC'):")
for region in shuffled_regions[:10]:
    print(f"  {region} -> sorted: {''.join(sorted(region))}")
```
