```python
# Random sampling from uniform probability matrix
import poolparty as pp
import pandas as pd

# 4-position motif with equal probabilities
prob_df = pd.DataFrame({'A': [0.25]*4, 'C': [0.25]*4, 'G': [0.25]*4, 'T': [0.25]*4})
with pp.Party() as party:
    pool = pp.from_prob_motif(prob_df, mode='random').named('uniform')
df = pool.generate_seqs(num_seqs=10, seed=42)
df
```

```python
# Biased motif: A-rich at position 0, T-rich at position 1
prob_df = pd.DataFrame({
    'A': [0.9, 0.1], 'C': [0.03, 0.1], 'G': [0.03, 0.1], 'T': [0.04, 0.7]
})
with pp.Party() as party:
    pool = pp.from_prob_motif(prob_df, mode='random').named('biased')
df = pool.generate_seqs(num_seqs=20, seed=42)
df
```

```python
# Joining with fixed flanks
with pp.Party() as party:
    left = pp.from_seq('ATG').named('left')
    var = pp.from_prob_motif(prob_df, mode='random').named('var')
    right = pp.from_seq('TAA').named('right')
    construct = pp.join([left, var, right]).named('construct')
df = construct.generate_seqs(num_seqs=10, seed=42)
df[['seq', 'var.seq']]
```
