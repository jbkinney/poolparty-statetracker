```python
import poolparty as pp; pp.init()

bg_seq = f"GCGCAATCGA<orf>ATGGCTAGCTACGAATGCAAGTACCCGTAA</orf>TTACGCGATCG"

blue_styles = ["lightblue", "deepskyblue", "steelblue"]
gold_styles = ["gold", "darkgoldenrod", "saddlebrown"]
frame_styles = blue_styles + gold_styles
codon_styles = ["salmon lower", "sienna upper"]

print("=== Codon styling (alternating red/blue for each codon) ===")
pp.from_seq(bg_seq).stylize_orf(region="orf", style_codons=codon_styles).named("codon_demo").print_library()

print("\n=== Frame styling (frame 0=red, 1=green, 2=blue) ===")
pp.from_seq(bg_seq).stylize_orf(region="orf", style_frames=frame_styles).named("frame_demo").print_library()

print("\n=== Frame styling with frame=2 (ORF starts at frame 2) ===")
pp.from_seq(bg_seq).stylize_orf(region="orf", style_frames=frame_styles, frame=2).named("frame_offset").print_library()

print("\n=== Reverse strand codon styling (frame=-1) ===")
pp.from_seq(bg_seq).stylize_orf(region="orf", style_codons=codon_styles, frame=-1).named("reverse_demo").print_library()

print("\n=== Frame styling with gaps (gaps skipped) ===")
bg_with_gap = f"GCGCAATCGA<orf_gap>ATGGCT---AGCTACGAATGCAAGTACCCGTAA</orf_gap>TTACGCGATCG"
pp.from_seq(bg_with_gap).stylize_orf(region="orf_gap", style_frames=frame_styles).named("gaps_demo").print_library()
```
