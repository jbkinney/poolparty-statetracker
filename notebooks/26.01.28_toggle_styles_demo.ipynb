```python
import poolparty as pp; pp.init()

# Demo: toggle_styles() for performance optimization
# When styles are disabled, Seq.style is None, avoiding all style overhead

# === With styles enabled (default) ===
print("=== With styles enabled (default) ===")
pool = pp.from_seqs(['ACGT', 'TTTT'], mode='sequential', style='red')
pool.print_library()

# === Disable styles for performance ===
print("\n=== Styles disabled (suppress_styles=True) ===")
pp.toggle_styles(on=False)
pool = pp.from_seqs(['ACGT', 'TTTT'], mode='sequential', style='red')
pool.print_library()

# stylize() is a no-op when styles suppressed
pool2 = pp.from_seq('GGGGCCCC').stylize(style='blue')
pool2.print_library()

# Complex operations work with suppressed styles
pool3 = pp.from_seq('ACGTACGT').mutagenize(num_mutations=2, style='yellow', mode='sequential')
pool3.print_library()

# === Re-enable styles ===
print("\n=== Styles re-enabled ===")
pp.toggle_styles(on=True)
pool4 = pp.from_seqs(['ACGT', 'TTTT'], mode='sequential', style='green')
pool4.print_library()
```
