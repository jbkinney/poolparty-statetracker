```python
import poolparty as pp
```

```python
# Basic mutagenize_scan with num_mutations in random mode
with pp.Party():
    seq = 'ACGTACGTACGT'
    result = pp.mutagenize_scan(seq, mutagenize_length=4, num_mutations=1, spacer_str='.', mode='random')
    print(f"Input: {seq}")
    print(f"Length: {len(seq)}, mutagenize_length: 4")
    print(f"num_states: {result.num_states}")
    df = pp.generate_library(result, seed=42, num_seqs=5)
    for s in df['seq']:
        print(f"  {s}")
```

```python
# mutagenize_scan with sequential mode (scans through all positions)
with pp.Party():
    seq = 'ACGTACGT'
    result = pp.mutagenize_scan(seq, mutagenize_length=3, num_mutations=1, spacer_str='.', mode='sequential', mark_changes=True)
    print(f"Input: {seq}")
    print(f"num_states (positions): {result.num_states}")
    df = pp.generate_library(result, seed=42, num_seqs=6)
    for s in df['seq']:
        print(f"  {s}")
```

```python
# mutagenize_scan with mutation_rate instead of num_mutations
with pp.Party():
    seq = 'ACGTACGTACGT'
    result = pp.mutagenize_scan(seq, mutagenize_length=6, mutation_rate=0.5, spacer_str='.', mark_changes=True, mode='sequential')
    print(f"Input: {seq}")
    print(f"mutation_rate=0.5 on window of length 6")
    df = pp.generate_library(result, seed=123, num_seqs=5)
    for s in df['seq']:
        print(f"  {s}")
```

```python
# mutagenize_scan with mark_changes=True to highlight mutations
with pp.Party():
    seq = 'ACGTACGTACGT'
    result = pp.mutagenize_scan(seq, mutagenize_length=4, num_mutations=2, spacer_str='.', mark_changes=True, mode='sequential')
    print(f"Input: {seq}")
    print(f"mark_changes=True: mutations shown in lowercase")
    df = pp.generate_library(result, seed=42, num_seqs=9)
    for s in df['seq']:
        print(f"  {s}")
```

```python
# mutagenize_scan with specific positions
with pp.Party():
    seq = 'ACGTACGTACGT'
    result = pp.mutagenize_scan(
        seq, 
        mutagenize_length=4, 
        num_mutations=1, 
        positions=[0, 4, 8],  # Only scan at these starting positions
        mode='sequential',
        mark_changes=True
    )
    print(f"Input: {seq}")
    print(f"Scanning positions: [0, 4, 8]")
    print(f"num_states: {result.num_states}")
    df = pp.generate_library(result, seed=42, num_seqs=3)
    for s in df['seq']:
        print(f"  {s}")
```
