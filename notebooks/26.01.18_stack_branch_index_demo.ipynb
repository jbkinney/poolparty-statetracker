```python
# Demo: Stack operation now tracks branch index in op[X]:stack.value
import poolparty as pp
pp.init()

# Create three pools with different numbers of states
pool_a = pp.from_seqs(['A1', 'A2', 'A3'], mode='sequential').named('A')  # 3 states
pool_b = pp.from_seqs(['B1', 'B2'], mode='sequential').named('B')        # 2 states
pool_c = pp.from_seqs(['C1', 'C2', 'C3', 'C4'], mode='sequential').named('C')  # 4 states

# Stack them together (9 total states)
stacked = pp.stack([pool_a, pool_b, pool_c]).named('stacked')

# Generate library with design cards
df = stacked.generate_library(num_cycles=1, report_design_cards=True)

# Show the relevant columns: stack.value now shows branch index (0, 1, 2)
# - Rows 0-2: branch 0 (pool A)
# - Rows 3-4: branch 1 (pool B)  
# - Rows 5-8: branch 2 (pool C)
stack_state_col = [c for c in df.columns if 'stack.value' in c][0]
cols_to_show = ['seq', 'stacked.value', 'A.value', 'B.value', 'C.value', stack_state_col]
tmp_df = df[cols_to_show].rename(columns={stack_state_col: 'branch_index'})
display(tmp_df)
```
