```python
import poolparty as pp; pp.init()


# Filtering + materialization - create a pool with known num_states from filtered results
print("\n=== Filter + Materialize (e.g., barcode generation) ===")
all_kmers = pp.get_kmers(6, mode="sequential")  # 4^6 = 4096 possible 6-mers
print(f"All 6-mers num_states: {all_kmers.num_states}")

# Filter to keep only those with balanced GC content (40-60%)
def balanced_gc(seq):
    gc = (seq.count('G') + seq.count('C')) / len(seq)
    return 0.4 <= gc <= 0.6

filtered = all_kmers.filter(balanced_gc)
print(f"Filtered pool num_states: {filtered.num_states}")  # None - unknown until generation

# Materialize exactly 10 balanced barcodes
barcodes = filtered.materialize(num_seqs=10, seed=42)
print(f"Materialized barcodes num_states: {barcodes.num_states}")  # Exactly 10
barcodes.print_library()

# Chaining after materialize - materialized pool can be used like any other pool
print("\n=== Chaining After Materialize ===")
mutated = barcodes.mutagenize(num_mutations=1, mode="random")
mutated.named("mutated_barcodes").print_library(num_seqs=5, seed=1)
```
