```python
import poolparty as pp; pp.init()

orf_seq = "ATGGCTAGCTACGAATGCAAGTACCCGTAA"
codon_style = ["gold", "cyan", "magenta"]

for frame in [1, 2, 3, -1, -2, -3]:
    print(f"=== frame={frame} ===")
    pp.from_seq(orf_seq).stylize_orf(style_codons=codon_style, frame=frame).named(f"frame{frame}").print_library()
    pp.from_seq(orf_seq).mutagenize_orf(num_mutations=1, mode="sequential", style="red bold", frame=frame).named(f"frame{frame}_mut").print_library(num_seqs=3)
```
