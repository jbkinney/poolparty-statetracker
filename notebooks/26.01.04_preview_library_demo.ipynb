```python
import poolparty as pp
```

```python
# Default behavior: prints 5 sequences when no args provided
with pp.Party():
    pool = pp.from_seqs(['ACGT', 'TGCA', 'GGGG'])
    pool.preview_library()
```

```python
# Specify num_seqs explicitly
with pp.Party():
    pool = pp.from_seqs(['ACGT', 'TGCA', 'GGGG'])
    pool.preview_library(num_seqs=2)
```

```python
# Specify num_cycles to iterate through all states
with pp.Party():
    pool = pp.from_seqs(['ACGT', 'TGCA', 'GGGG'])
    pool.preview_library(num_cycles=2)
```
