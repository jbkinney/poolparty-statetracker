```python
import poolparty as pp; pp.init()
import pandas as pd

print("Loading E. coli coordinate data and genome FASTA")
df = pd.read_csv("26.02.02_ecoli_library/20260202_df_peaks_collapsed_14500.csv")
fasta_path = "26.02.02_ecoli_library/ecoli_geome_U00096.3.fasta"
print(f"Loaded {len(df)} coordinates")

print("\nExtracting sequences from genome using from_fasta()")
strand_map = {"forward": "+", "reverse": "-"}
chrom = "U00096.3"
coordinates = [
    (chrom, start, stop, strand_map[strand])
    for start, stop, strand in zip(
        df["tss_window_sequence_start"],
        df["tss_window_sequence_end"],
        df["strand"]
    )
]

fixed_seq_1 = 'AAGACTCAACCAATGACCCTCGAG'
fixed_seq_2 = 'GCTAGCCTATGTGTCCTACGAAGA'

pool = pp.from_fasta(fasta_path, coordinates, prefix='ecoli', style='cyan')

oligos = pp.join([fixed_seq_1, pool, fixed_seq_2])

#short_pool = pool.slice_seq(start=0, stop=50)
pool[-5:].print_library()
```
