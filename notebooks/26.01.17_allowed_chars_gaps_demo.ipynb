```python
import poolparty as pp; pp.init()
pp.add_highlight(which='lower', style='bold cyan')

# Use / as a visual separator (e.g., for splice sites)
# allowed_chars can include / to align with sequence positions
wt_pool = pp.from_seq('acgtAGGA|GTATAGacgt')
mut1_pool = pp.mutagenize(wt_pool, 
                          num_mutations=1, 
                          allowed_chars='acgtNNNN|GYNNNNacgt',  # G or C/T at splice site
                          mode='sequential').named('mut1')
mut2_pool = pp.mutagenize(wt_pool, 
                          num_mutations=2, 
                          allowed_chars='acgtNNNN|GYNNNNacgt', 
                          mode='sequential').named('mut2')

combo_pool = pp.stack([
    wt_pool.repeat_states(10), 
    mut1_pool.repeat_states(3), 
    mut2_pool.repeat_states(1)
]).print_library()
```
