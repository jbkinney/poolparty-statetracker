```python
import poolparty as pp
from typing import Literal

pp.init()
pp.toggle_cards(on=True)
seq_pool = pp.from_seq('ACGTACGT')
pool = seq_pool.shuffle_seq()
pool = seq_pool.mutagenize(num_mutations=2, mode='random', style='bold yellow')
pool.print_library(num_seqs=10)

pool.print_dag
```
