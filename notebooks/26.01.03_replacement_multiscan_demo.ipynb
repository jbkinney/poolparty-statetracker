```python
import poolparty as pp
from poolparty import replacement_multiscan
```

```python
# Basic usage: replace 3-char segments at 2 random positions with a single Pool (deepcopied)
with pp.Party() as party:
    bg = pp.from_seqs(['ACGTACGTACGTACGTACGT'])  # 20 chars
    ins = pp.from_seq('GGG')  # 3-char replacement
    result = replacement_multiscan(bg, num_replacements=2, replacement_pools=ins, spacer_str='.').named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# Using multiple distinct pools for replacements
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    pool1 = pp.from_seq('GGG')
    pool2 = pp.from_seq('TTT')
    result = replacement_multiscan(
        bg, num_replacements=2, replacement_pools=[pool1, pool2], spacer_str='.', 
    ).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# Using hybrid mode with fixed number of states
with pp.Party() as party:
    bg = pp.from_seqs(['ACGTACGTACGTACGTACGT'])  # 20 chars
    ins = pp.from_seq('XXX')
    result = replacement_multiscan(
        bg, num_replacements=2, replacement_pools=ins,
        mode='hybrid', num_hybrid_states=5, spacer_str='.'
    ).named('result')
result.generate_library(num_seqs=10, seed=42)['seq']
```

```python
# Three replacements with mark_changes=True (swapcase)
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    ins = pp.from_seq('GGG')
    result = replacement_multiscan(
        bg, num_replacements=3, replacement_pools=ins, mark_changes=True
    ).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# Using spacer_str to mark replacement boundaries
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    ins = pp.from_seq('G')  # 1-char replacement
    result = replacement_multiscan(
        bg, num_replacements=2, replacement_pools=ins, spacer_str='.'
    ).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# insertion_mode='ordered' (default): pool1 always goes to leftmost position, pool2 to next, etc.
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    pool1 = pp.from_seq('GGG')  # Always appears at leftmost selected position
    pool2 = pp.from_seq('TTT')  # Always appears at rightmost selected position
    result = replacement_multiscan(
        bg, num_replacements=2, replacement_pools=[pool1, pool2],
        spacer_str='.', insertion_mode='ordered'
    ).named('result')
df = result.generate_library(num_seqs=5, seed=42)
print("insertion_mode='ordered': GGG always left of TTT")
df['seq']
```

```python
# insertion_mode='unordered': pools are randomly assigned to positions
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    pool1 = pp.from_seq('GGG')  # Can appear at either position
    pool2 = pp.from_seq('TTT')  # Can appear at either position
    result = replacement_multiscan(
        bg, num_replacements=2, replacement_pools=[pool1, pool2],
        spacer_str='.', insertion_mode='unordered'
    ).named('result')
df = result.generate_library(num_seqs=10, seed=42)
print("insertion_mode='unordered': GGG and TTT can appear in either order")
df['seq']
```
