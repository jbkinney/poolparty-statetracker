```python
import poolparty as pp
```

```python
# Uppercase IUPAC input -> uppercase output
with pp.Party() as party:
    iupac_upper = 'GGA.NNNNN.CAG'
    pool = pp.from_iupac_motif(iupac_upper, mark_changes=False).named('iupac')
df = pool.generate_library(num_seqs=5, seed=42)
print(f"Input: {iupac_upper}")
print(df[['seq']])
```

```python
# Lowercase IUPAC input -> lowercase output (case preserved)
with pp.Party() as party:
    iupac_lower = 'gga.nnnnn.cag'
    pool = pp.from_iupac_motif(iupac_lower, mark_changes=False).named('iupac')
df = pool.generate_library(num_seqs=5, seed=42)
print(f"Input: {iupac_lower}")
print(df[['seq']])
```

```python
# Mixed case input -> mixed case output (case preserved per position)
with pp.Party() as party:
    iupac_mixed = 'GGA.nnnnn.CAG'
    pool = pp.from_iupac_motif(iupac_mixed, mark_changes=False).named('iupac')
df = pool.generate_library(num_seqs=5, seed=42)
print(f"Input: {iupac_mixed}")
print(df[['seq']])
```

```python
# mark_changes=True swaps case at degenerate positions for visualization
with pp.Party() as party:
    iupac_mixed = 'GGA.nnnnn.CAG'
    pool = pp.from_iupac_motif(iupac_mixed, mark_changes=False).named('iupac')
df = pool.generate_library(num_seqs=5, seed=42)
print(f"Input: {iupac_mixed} (mark_changes=True)")
print(df[['seq']])
```
