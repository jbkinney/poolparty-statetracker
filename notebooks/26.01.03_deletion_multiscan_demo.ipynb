```python
import poolparty as pp
```

```python
# Basic usage: delete 3-char segments at 2 random positions
with pp.Party() as party:
    bg = pp.from_seqs(['ACGTACGTACGTACGTACGT'])  # 20 chars
    result = pp.deletion_multiscan(bg, deletion_length=3, num_deletions=2).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# String input also works
with pp.Party() as party:
    result = pp.deletion_multiscan('ACGTACGTACGTACGTACGT', deletion_length=3, num_deletions=2).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# Without marker: segments are removed entirely
with pp.Party() as party:
    seq = 'ACGTACGTACGTACGTACGT'
    result = pp.deletion_multiscan(
        seq,
        deletion_length=3,
        num_deletions=2,
        deletion_marker=None
    ).named('result')
print(seq, '# original')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# Custom marker character
with pp.Party() as party:
    result = pp.deletion_multiscan(
        'ACGTACGTACGTACGTACGT',
        deletion_length=3,
        num_deletions=2,
        deletion_marker='X'
    ).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# Using spacer strings around deletion sites
with pp.Party() as party:
    result = pp.deletion_multiscan(
        'ACGTACGTACGTACGTACGT',
        deletion_length=3,
        num_deletions=2,
        spacer_str='.'
    ).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```

```python
# Hybrid mode with fixed number of states
with pp.Party() as party:
    result = pp.deletion_multiscan(
        'ACGTACGTACGTACGTACGT',
        deletion_length=3,
        num_deletions=2,
        mode='hybrid',
        num_hybrid_states=5
    ).named('result')
result.generate_library(num_seqs=15, seed=42)['seq']
```

```python
# Restrict deletion positions using positions parameter
with pp.Party() as party:
    result = pp.deletion_multiscan(
        'ACGTACGTACGTACGTACGT',
        deletion_length=3,
        num_deletions=2,
        positions=[0, 5, 10, 15]  # Only these starting positions
    ).named('result')
result.generate_library(num_seqs=10, seed=42)['seq']
```

```python
# Using slice for positions
with pp.Party() as party:
    result = pp.deletion_multiscan(
        'ACGTACGTACGTACGTACGT',
        deletion_length=3,
        num_deletions=2,
        positions=slice(0, 10)  # Only first half of sequence
    ).named('result')
result.generate_library(num_seqs=10, seed=42)['seq']
```

```python
# Three simultaneous deletions
with pp.Party() as party:
    result = pp.deletion_multiscan(
        'ACGTACGTACGTACGTACGTACGTACGT',  # 28 chars
        deletion_length=4,
        num_deletions=3
    ).named('result')
result.generate_library(num_seqs=5, seed=42)['seq']
```
