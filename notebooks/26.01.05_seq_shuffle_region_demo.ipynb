```python
import poolparty as pp; pp.init()

# seq_shuffle with region=[start, stop] interval
pool = pp.seq_shuffle('AAAAACGTACGT', region=[5, 9])
df = pool.generate_library(num_seqs=5, seed=42)
print("region=[5, 9] shuffles positions 5-8, preserving AAAAA and ACGT:")
print(df['seq'].tolist())

# seq_shuffle with region=None (entire sequence)
pool2 = pp.seq_shuffle('ABCD')
df2 = pool2.generate_library(num_seqs=3, seed=123)
print("\nregion=None shuffles entire sequence:")
print(df2['seq'].tolist())
```

```python
import poolparty as pp; pp.init()

# seq_shuffle with region=marker_name (marker-based)
pool = pp.seq_shuffle('AAA<r>BCDE</r>FFF', region='r')
df = pool.generate_library(num_seqs=5, seed=42)
print("region='r' shuffles marker content, preserving AAA, FFF, and marker tags:")
print(df['seq'].tolist())

# Using OpsContainer.seq_shuffle() with marker
bg = pp.from_seq('XX<target>1234</target>YY')
result = bg.ops.seq_shuffle('target', remove_marker=True)
df2 = result.generate_library(num_seqs=4, seed=99)
print("\nOpsContainer.seq_shuffle with remove_marker=True:")
print(df2['seq'].tolist())
```
