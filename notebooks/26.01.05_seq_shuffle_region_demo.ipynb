```python
import poolparty as pp; pp.init()

# shuffle_seq with region=[start, stop] interval, mark_changes=False (preserve case)
pool = pp.shuffle_seq('AAAAACGTACGT', region=[5, 9], mark_changes=False)
df = pool.generate_library(num_seqs=5, seed=42)
print("region=[5, 9], mark_changes=False (preserves case):")
print(df['seq'].tolist())

# shuffle_seq with mark_changes=True (swap case of shuffled region)
pool2 = pp.shuffle_seq('AAAAACGTACGT', region=[5, 9], mark_changes=True)
df2 = pool2.generate_library(num_seqs=3, seed=42)
print("\nregion=[5, 9], mark_changes=True (swaps case):")
print(df2['seq'].tolist())
```

```python
import poolparty as pp; pp.init()

# shuffle_seq with region=marker_name (marker-based), mark_changes=False
pool = pp.shuffle_seq('AAA<r>BCDE</r>FFF', region='r', mark_changes=False)
df = pool.generate_library(num_seqs=5, seed=42)
print("region='r' (marker-based), mark_changes=False:")
print(df['seq'].tolist())

# OpsContainer.shuffle_seq with marker and mark_changes=True
bg = pp.from_seq('XX<target>ACGT</target>YY')
result = bg.ops.shuffle_seq('target', remove_marker=True, mark_changes=True)
df2 = result.generate_library(num_seqs=4, seed=99)
print("\nOpsContainer.shuffle_seq, mark_changes=True, remove_marker=True:")
print(df2['seq'].tolist())
```
