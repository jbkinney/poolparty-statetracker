```python
import poolparty as pp; pp.init()

# Basic usage: restrict mutations to purines only (R=A,G)
# Sequence must contain only allowed chars at each position
pp.mutagenize('AGAG', num_mutations=1, allowed_chars='RRRR', mode='sequential').print_library()

# Position-specific constraints using different IUPAC codes
# R=purines(A,G), Y=pyrimidines(C,T), S=(G,C), W=(A,T)
pp.mutagenize('ACGT', num_mutations=1, allowed_chars='RYSW', mode='sequential').print_library()

# Skip positions with no valid mutations (A can only be A -> no mutation)
pp.mutagenize('AC', num_mutations=1, allowed_chars='AN', mode='sequential').print_library()

# Double mutations with constrained alphabet
pp.mutagenize('AG', num_mutations=2, allowed_chars='RR', mode='sequential').print_library()

# Random mode with allowed_chars
pp.mutagenize('AGAGAGAG', mutation_rate=0.3, allowed_chars='RRRRRRRR').generate_library(num_seqs=5, seed=42)
```
