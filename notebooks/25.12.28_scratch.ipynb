```python
import poolparty as pp
bg_L = 20
ins_L = 5
with pp.Party() as party:
    bg = pp.from_seqs(['A'*bg_L, 'T'*bg_L]).named('bg').iteration_order(0)
    ins = pp.get_kmers(length=ins_L, mode='hybrid', hybrid_mode_num_states=2).named('insert').iteration_order(1)
    res = pp.replacement_scan(bg_pool=bg, ins_pool=ins, spacer_str='.', mode='hybrid', hybrid_mode_num_states=2, iteration_order=2)
res.print_tree()
pp.generate_seqs(pool=res, num_complete_iterations=1, seed=3)['seq']
```
