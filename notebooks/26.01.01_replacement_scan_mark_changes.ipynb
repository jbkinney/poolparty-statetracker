```python
import poolparty as pp
```

```python
# swapcase() swaps the case of all characters in a sequence
pool = pp.swapcase('ACGT').named('swapped')
pool.generate_library(num_seqs=1)
```

```python
# replacement_scan with mark_changes=True highlights the insert by swapping its case
bg = 'AAAAAAAAAA'
ins = 'TTT'
pool = pp.replacement_scan(bg, ins, mark_changes=True, mode='sequential').named('scan')
pool.generate_library(num_cycles=1)
```

```python
# Party default for mark_changes can be set via set_default()
with pp.Party() as party:
    party.set_default('mark_changes', True)
    pool = pp.replacement_scan('AAAAAAAAAA', 'GGG', positions=[0], mode='sequential').named('result')
    df = pool.generate_library(num_cycles=1)
df
```
