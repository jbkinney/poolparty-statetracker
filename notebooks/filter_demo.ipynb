```python
import poolparty as pp; pp.init()

# Create a source pool with 10 sequences
root = pp.from_seqs(
    ["AAAA", "CCCC", "GGGG", "TTTT", "ACGT", 
     "TGCA", "ATAT", "CGCG", "GGCC", "AAGG"],
    mode="sequential"
)

# Filter to sequences starting with 'A' or 'G'
filtered = root.filter(lambda s: s[0] in "AG")

# Default: shows all rows including NullSeq as None
print("=== Default (shows null rows) ===")
filtered.print_library(num_seqs=10)

# With discard_null_seqs=True: only valid sequences
print("\n=== With discard_null_seqs=True ===")
filtered.print_library(num_seqs=5, discard_null_seqs=True)

# Chained filters
print("\n=== Chained filters: starts with A AND ends with T ===")
chained = root.filter(lambda s: s.startswith("A")).filter(lambda s: s.endswith("T"))
chained.print_library(num_seqs=2, discard_null_seqs=True)

# Filter with downstream operation
print("\n=== Filter + downstream operation ===")
result = filtered.upper()  # NullSeq propagates through
result.print_library(num_seqs=5, discard_null_seqs=True)
```
