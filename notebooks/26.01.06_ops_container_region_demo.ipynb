```python
import poolparty as pp; pp.init()
import pandas as pd

# Create a pool with a marked region
print('base pool')
pool = pp.from_seq('AAAA<reg>ACAT</reg>CCCC').print_library()

# Base ops via OpsContainer (use native region support)
print('mutagenize')
pool.mutagenize('reg', num_mutations=1, mode='sequential').print_library()
print('shuffle_seq')
pool.shuffle_seq('reg').print_library()

# Scan ops via OpsContainer (use native region support)
print('deletion_scan')
pool.deletion_scan('reg', deletion_length=2, mode='sequential').print_library()
print('insertion_scan')
pool.insertion_scan('reg', 'GG', mode='sequential').print_library()
print('replacement_scan')
pool.replacement_scan('reg', 'GG', mode='sequential').print_library()
print('mutagenize_scan')
pool.mutagenize_scan('reg', mutagenize_length=2, num_mutations=1, mode='sequential').print_library()
print('shuffle_scan')
pool.shuffle_scan('reg', shuffle_length=2, mode='sequential').print_library()

# Insert operations via OpsContainer (use native bg_pool/region support)
print('insert_from_iupac')
pool.insert_from_iupac('reg', 'RY', mode='sequential').print_library()
print('insert_from_motif')
prob_df = pd.DataFrame({'A': [0.5, 0.5], 'C': [0.5, 0.5], 'G': [0, 0], 'T': [0, 0]})
pool.insert_from_motif('reg', prob_df).print_library()
print('insert_kmers')
pool.insert_kmers('reg', length=2, mode='sequential').print_library()

# Demonstrate remove_marker=False preserves markers
print('mutagenize_remove_marker_true')
pool.mutagenize('reg', num_mutations=1, remove_marker=True, mode='sequential').print_library()
```
