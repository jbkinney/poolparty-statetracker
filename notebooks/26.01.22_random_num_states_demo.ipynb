```python
import poolparty as pp; pp.init()

# Demonstrate random mode behavior with auto-sync to parent states
print("=== Auto-sync behavior ===\n")

# When an operation has mode='random' and no explicit num_states, 
# it automatically syncs to the parent pool's state
parent = pp.from_seqs(['AAAA', 'TTTT', 'GGGG'], mode='sequential')  # 3 states
child = pp.mutagenize(parent, num_mutations=1, mode='random').named('mutant')
print(f"Parent has {parent.num_states} states")
print(f"Child (auto-synced) has {child.num_states} states")
child.print_library()

# The same state always produces the same random output (reproducible)
print("=== Reproducibility (same output in each cycle) ===")
child.print_library(num_cycles=2)

# When explicit num_states is provided, it overrides auto-sync
print("\n=== Explicit num_states ===")
pool2 = pp.mutagenize('ACGT', num_mutations=1, mode='random', num_states=5).named('mutant')
print(f"Pool with explicit num_states=5 has {pool2.num_states} states")
pool2.print_library()

# Source operations (no parents) with mode='random' and no num_states are stateless
print("\n=== Stateless source operations ===")
pool3 = pp.get_kmers(length=3, mode='random')
print(f"get_kmers without num_states: num_states={pool3.num_states}")
# Must use num_seqs since there's no state to iterate
pool3.named('kmer').print_library(num_seqs=5)

# With explicit num_states, source operations have reproducible random states
print("\n=== Source with explicit num_states ===")
pool4 = pp.get_kmers(length=3, mode='random', num_states=5).named('kmer')
print(f"get_kmers with num_states=5: num_states={pool4.num_states}")
pool4.print_library()
```
