```python
import poolparty as pp; pp.init()

# Demonstrate random mode with num_states parameter
# When num_states=None (default), random mode behaves like before (num_values=1)
pool1 = pp.mutagenize('ACGT', num_mutations=1, mode='random').print_library()

# When num_states is provided, random mode cycles through that many states
# Each state produces random output, but the same state always produces the same output with the same seed
pool2 = pp.mutagenize('ACGT', num_mutations=1, mode='random', num_states=5).print_library()

# Works with other operations too
pool3 = pp.get_kmers(length=3, mode='random', num_states=10).print_library()

pool4 = pp.from_seqs(['AAA', 'TTT', 'GGG'], mode='random', num_states=8).print_library()
```
