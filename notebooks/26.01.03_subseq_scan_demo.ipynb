```python
import poolparty as pp
```

```python
# Basic subseq_scan: extract 4-bp subsequences at all positions (sequential mode)
with pp.Party():
    result = pp.subseq_scan('ACGTACGTACGT', seq_length=4, mode='sequential').named('result')

df = result.generate_library(num_cycles=1, seed=42)
df
```

```python
# subseq_scan with specific positions
with pp.Party():
    result = pp.subseq_scan(
        'ACGTacgtACGT',
        seq_length=4,
        positions=[0, 4, 8],
        mode='sequential'
    ).named('result')

df = result.generate_library(num_cycles=2, seed=42)
df
```

```python
# subseq_scan with strand='-' (reverse complement of extracted region)
with pp.Party():
    result = pp.subseq_scan(
        'ACGTACGTACGT',
        seq_length=4,
        positions=[0],
        strand='-',
        mode='sequential'
    ).named('result')

df = result.generate_library(num_seqs=5, seed=42)
print("Original: ACGT (first 4 bp)")
print("Reverse complement: ACGT")
df
```

```python
# subseq_scan with strand='both' - scans both strands (2x states)
with pp.Party():
    result = pp.subseq_scan(
        'ACGCTTTT',
        seq_length=4,
        positions=[0,2,4],
        strand='both',
        mode='sequential'
    ).named('result')

df = result.generate_library(num_seqs=10, seed=42)
print("Position 0: ACGT on + strand, reverse complement on - strand")
df
```

```python
# subseq_scan with random mode (default) - samples random positions
with pp.Party():
    result = pp.subseq_scan('ACGTACGTACGT', seq_length=4).named('result')

df = result.generate_library(num_seqs=10, seed=42)
print("Random 4-bp subsequences from ACGTACGTACGT:")
df
```
