```python
import poolparty as pp
bg_L = 20
ins_L = 5
with pp.Party() as party:
    bg = pp.from_seqs(['A'*bg_L, 'T'*bg_L], 
                      op_iter_order=-1, 
                      mode='sequential',
                      name='bg')
    
    print(f'{bg.operation.counter.iter_order=}')
    print(f'{bg.counter.iter_order=}')
    
    mut = pp.mutagenize(pool=bg, 
                        num_mutations=3,
                        mode='hybrid',
                        op_iter_order=-1,
                        num_hybrid_states=4,
                        name='mut')
    
    print(f'{mut.operation.counter.iter_order=}')
    print(f'{mut.counter.iter_order=}')
    
    result =pp.state_shuffle(mut)
    
    print(f'{result.operation.counter.iter_order=}')
    print(f'{result.counter.iter_order=}')
    
result.counter.print_tree()

pp.generate_seqs(pool=result, num_complete_iterations=1, seed=3)['seq']
```
