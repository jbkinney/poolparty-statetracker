```python
import poolparty as pp; pp.init()
import tempfile, os

# Create a test FASTA file
fasta_content = """>chr1
ACGTACGTACGTACGTACGTACGT
>chr2
AAAACCCCGGGGTTTTAAAACCCC
"""
with tempfile.NamedTemporaryFile(mode='w', suffix='.fasta', delete=False) as f:
    f.write(fasta_content)
    fasta_path = f.name

# Basic extraction (0-based coordinates)
pool = pp.from_fasta(fasta_path, 'chr1', 0, 12).named('basic')
pool.print_library()

# Extract with reverse complement (strand='-')
pool_rc = pp.from_fasta(fasta_path, 'chr2', 0, 8, strand='-').named('revcomp')
print("\nReverse complement (AAAACCCC -> GGGGTTTT):")
pool_rc.print_library()

# Use extracted sequence with downstream operations
pool_mutated = (
    pp.from_fasta(fasta_path, 'chr1', 4, 16)
    .deletion_scan(deletion_length=2)
    .named('mutated')
)
print("\nDeletion scan on extracted region:")
pool_mutated.print_library()

# Cleanup
os.unlink(fasta_path)
if os.path.exists(fasta_path + '.fai'):
    os.unlink(fasta_path + '.fai')
```
