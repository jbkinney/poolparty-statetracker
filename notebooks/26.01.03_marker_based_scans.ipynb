```python
# Marker-based scan architecture demo
# insertion_scan, replacement_scan, and deletion_scan now use marker_scan internally

import poolparty as pp
```

```python
# insertion_scan: inserts a sequence at scanning positions
# Internally uses marker_scan (zero-length marker) + replace_marker_content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    ins = pp.from_seq('GGG')
    result = pp.insertion_scan(bg, ins, mode='sequential').named('result')

df = result.generate_seqs(num_complete_iterations=1)
df
```

```python
# replacement_scan: replaces a segment at scanning positions
# Internally uses marker_scan (region marker) + replace_marker_content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    ins = pp.from_seq('GGG')
    result = pp.replacement_scan(bg, ins, mode='sequential').named('result')

df = result.generate_seqs(num_complete_iterations=1)
df
```

```python
# deletion_scan: deletes a segment at scanning positions
# Internally uses marker_scan (region marker) + replace_marker_content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    result = pp.deletion_scan(bg, deletion_length=3, mode='sequential').named('result')

df = result.generate_seqs(num_complete_iterations=1)
df
```

```python
# deletion_scan with deletion_marker=None removes segment entirely

with pp.Party() as party:
    bg = pp.from_seq('ACGTACGTACGT')
    result = pp.deletion_scan(bg, deletion_length=3, deletion_marker=None, mode='sequential').named('result')

df = result.generate_seqs(num_complete_iterations=1)
df
```

```python
# spacer_str adds separators around the inserted/replaced content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    ins = pp.from_seq('GGG')
    result = pp.insertion_scan(bg, ins, spacer_str='.', mode='sequential').named('result')

df = result.generate_seqs(num_complete_iterations=1)
df
```

```python
# The underlying marker_scan operation inserts XML markers at positions
# This is what powers insertion_scan, replacement_scan, and deletion_scan

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    marked = pp.marker_scan(bg, marker='region', marker_length=3, mode='sequential').named('marked')

df = marked.generate_seqs(num_complete_iterations=1)
df
```

```python
# apply_at_marker transforms the content at a marked region
# This enables powerful composable operations

with pp.Party() as party:
    bg = pp.from_seq('ACGT<orf>ATGCCC</orf>TTTT')
    result = pp.apply_at_marker(bg, 'orf', pp.reverse_complement).named('result')

df = result.generate_seqs(num_seqs=1)
print(f"Original: ACGT<orf>ATGCCC</orf>TTTT")
print(f"Result:   {df['seq'].iloc[0]}")
```
