```python
# Marker-based scan architecture demo
# insertion_scan, replacement_scan, and deletion_scan now use marker_scan internally

import poolparty as pp
```

```python
# insertion_scan: inserts a sequence at scanning positions
# Internally uses marker_scan (zero-length marker) + replace_marker_content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    ins = pp.from_seq('GGG')
    result = pp.insertion_scan(bg, ins, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# replacement_scan: replaces a segment at scanning positions
# Internally uses marker_scan (region marker) + replace_marker_content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    ins = pp.from_seq('GGG')
    result = pp.replacement_scan(bg, ins, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# deletion_scan: deletes a segment at scanning positions
# Internally uses marker_scan (region marker) + replace_marker_content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    result = pp.deletion_scan(bg, deletion_length=3, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# deletion_scan with deletion_marker=None removes segment entirely

with pp.Party() as party:
    bg = pp.from_seq('ACGTACGTACGT')
    result = pp.deletion_scan(bg, deletion_length=3, deletion_marker=None, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# spacer_str adds separators around the inserted/replaced content

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    ins = pp.from_seq('GGG')
    result = pp.insertion_scan(bg, ins, spacer_str='.', mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# The underlying marker_scan operation inserts XML markers at positions
# This is what powers insertion_scan, replacement_scan, and deletion_scan

with pp.Party() as party:
    bg = pp.from_seq('AAAAAAAAAA')
    marked = pp.marker_scan(bg, marker='region', marker_length=3, mode='sequential').named('marked')

df = marked.generate_seqs(num_cycles=1)
df
```

```python
# apply_at_marker transforms the content at a marked region
# This enables powerful composable operations

with pp.Party() as party:
    bg = pp.from_seq('ACGT<orf>ATGCCC</orf>TTTT')
    result = pp.apply_at_marker(bg, 'orf', pp.reverse_complement).named('result')

df = result.generate_seqs(num_seqs=1)
print(f"Original: ACGT<orf>ATGCCC</orf>TTTT")
print(f"Result:   {df['seq'].iloc[0]}")
```

# Unified `scan()` function

The three scan functions (`insertion_scan`, `replacement_scan`, `deletion_scan`) are now unified into a single `scan()` function with an `action` parameter.

```python
# scan() with action='insert' - equivalent to insertion_scan()
# Inserts a sequence at scanning positions (adds to background length)

with pp.Party() as party:
    result = pp.scan('insert', 'AAAAAAAAAA', ins_pool='GGG', mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# scan() with action='replace' - equivalent to replacement_scan()
# Replaces a segment at scanning positions (preserves background length)

with pp.Party() as party:
    result = pp.scan('replace', 'AAAAAAAAAA', ins_pool='GGG', mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# scan() with action='delete' and mark_changes=True
# Fills the deletion gap with del_char (default '-')

with pp.Party() as party:
    result = pp.scan('delete', 'AAAAAAAAAA', del_length=3, mark_changes=True, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# scan() with action='delete' and mark_changes=False
# Simply removes the segment (reduces length)

with pp.Party() as party:
    result = pp.scan('delete', 'AAAAAAAAAA', del_length=3, mark_changes=False, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# Custom del_char for deletion marker

with pp.Party() as party:
    result = pp.scan('delete', 'AAAAAAAAAA', del_length=3, del_char='X', 
                     mark_changes=True, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# mark_changes=True with 'replace' applies swap_case to highlight the insert

with pp.Party() as party:
    result = pp.scan('replace', 'AAAAAAAAAA', ins_pool='GGG', 
                     mark_changes=True, mode='sequential').named('result')

df = result.generate_seqs(num_cycles=1)
df
```

```python
# The wrapper functions (insertion_scan, replacement_scan, deletion_scan) 
# are now thin wrappers around scan() for backward compatibility

# These are equivalent:
#   pp.insertion_scan(bg, ins)          <==>  pp.scan('insert', bg, ins_pool=ins)
#   pp.replacement_scan(bg, ins)        <==>  pp.scan('replace', bg, ins_pool=ins)
#   pp.deletion_scan(bg, deletion_length=3)  <==>  pp.scan('delete', bg, del_length=3, mark_changes=True)
#   pp.deletion_scan(bg, deletion_length=3, deletion_marker=None)  <==>  pp.scan('delete', bg, del_length=3, mark_changes=False)

print("scan() API summary:")
print("  action='insert'  - requires ins_pool, adds to sequence length")
print("  action='replace' - requires ins_pool, preserves sequence length")
print("  action='delete'  - requires del_length, mark_changes controls gap fill")
```
