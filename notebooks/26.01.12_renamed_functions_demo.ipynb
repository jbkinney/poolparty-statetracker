```python
import poolparty as pp; pp.init()
import pandas as pd

# === from_iupac (formerly from_iupac_motif) ===
print('=== from_iupac ===')
pp.from_iupac('RYN', mode='sequential').named('iupac').print_library()

# === from_motif (formerly from_prob_motif) ===
print('\n=== from_motif ===')
prob_df = pd.DataFrame({'A': [1.0, 0.5], 'T': [0.0, 0.5]})
pool = pp.from_motif(prob_df, mode='random').named('motif')
seqs = pool.generate_library(num_seqs=4, seed=42, seqs_only=True)
for i, seq in enumerate(seqs): print(f'{i:5d}  {seq}')

# === slice_seq (formerly seq_slice) ===
print('\n=== slice_seq ===')
pp.from_seq('ACGTACGT').named('original').print_library(show_header=False)
pp.slice_seq(pp.from_seq('ACGTACGT'), slice(0, 4)).named('sliced').print_library()

# === clear_gaps (formerly clear_gap_chars) ===
print('\n=== clear_gaps ===')
pp.from_seq('A-C.G T').named('gapped').print_library(show_header=False)
pp.clear_gaps(pp.from_seq('A-C.G T')).named('cleaned').print_library()

# === rc (unchanged, for reference) ===
print('\n=== rc ===')
pp.rc(pp.from_seq('ACGT')).named('rc').print_library()
```
