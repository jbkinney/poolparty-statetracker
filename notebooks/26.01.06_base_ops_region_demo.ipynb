```python
import poolparty as pp; pp.init()
import pandas as pd

# Background sequence with a marker
bg = 'AAA<region>XXX</region>TTT'

# By default, pp.init() sets mark_changes=True, so inserts are swapcased (lowercase)
print('=== get_kmers with region (default mark_changes=True) ===')
pp.get_kmers(2, bg_pool=bg, region='region', mode='sequential').print_library()

# With mark_changes=False, inserts stay uppercase
print('\n=== get_kmers with mark_changes=False ===')
pp.get_kmers(2, bg_pool=bg, region='region', mark_changes=False, mode='sequential').print_library()

# from_iupac_motif with region (uses default mark_changes=True)
print('\n=== from_iupac_motif with region ===')
pp.from_iupac_motif('RY', bg_pool=bg, region='region', mode='sequential').print_library()

# from_prob_motif with region (uses default mark_changes=True)
print('\n=== from_prob_motif with region ===')
prob_df = pd.DataFrame({'A': [1, 0], 'C': [0, 1], 'G': [0, 0], 'T': [0, 0]})
pp.from_prob_motif(prob_df, bg_pool=bg, region='region').print_library(num_seqs=3, seed=42)

# from_seqs with region (uses default mark_changes=True)
print('\n=== from_seqs with region ===')
pp.from_seqs(['GG', 'CC'], bg_pool=bg, region='region', mode='sequential').print_library()

# remove_marker=True removes the marker tags from output
print('\n=== get_kmers with remove_marker=True ===')
pp.get_kmers(2, bg_pool=bg, region='region', remove_marker=True, mode='sequential').print_library()
```
