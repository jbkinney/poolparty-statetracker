```python
import poolparty as pp; pp.init()

print("Basic reverse translation (codon_selection='first')")
pp.from_seq("ATGCCC").translate().reverse_translate().print_library()

print("Reverse translation from protein string")
pp.reverse_translate("MAP").print_library()

print("Reverse translation with random codon selection")
pp.reverse_translate("MAP", codon_selection="random").print_library(num_seqs=5, seed=42)

print("Random selection with explicit num_states")
pp.reverse_translate("MAP", codon_selection="random", num_states=3).print_library()

print("Round-trip: translate then reverse_translate (synonymous DNA)")
original = pp.from_seq("ATGGCTCCC")
translated = original.translate()
back_to_dna = translated.reverse_translate()
print("Original DNA:")
original.print_library()
print("Translated protein:")
translated.print_library()
print("Reverse-translated DNA:")
back_to_dna.print_library()

print("Reverse translation with stop codon")
pp.from_seq("ATGGCTTGA").translate().reverse_translate().print_library()

print("Style preservation: DNA codon styles -> protein -> DNA")
protein = pp.from_seq("ATGGCTCCC").annotate_orf("cds", frame=1, style_codons=["red","green","blue"])\
    .translate(region="cds", preserve_codon_styles=True)
protein.print_library()
protein.reverse_translate().print_library()
```
