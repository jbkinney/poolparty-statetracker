```python
import poolparty as pp
L=5
with pp.Party() as party:
    A = pp.from_seqs(['A'*L,'C'*L, 'G'*L, 'T'*L], mode='sequential').named('A')
    B = pp.get_kmers(length=L, mode='hybrid', hybrid_mode_num_states=4).named('B')
    C = pp.get_kmers(length=L, mode='random').named('C')
    Ax2 = (2*A).named('Ax2')
    Bx2 = (B*2).named('Bx2')
    D = pp.concat([Ax2,Bx2]).named('D')
    E = pp.concat([C,D]).named('E')
    F = E.copy().named('F')
    pp.sync([E, F])
    G = pp.concat([E, F[::-1]]).named('G')

party.print_graph(style='clean')
```

```python
df = G.generate(num_complete_iterations=1, 
                report_seq=True,
                report_pool_seqs=False, 
                report_pool_states=True, 
                report_op_states=True,
                report_op_keys=True, 
                pools_to_report='all',
                organize_columns_by='type')
display(df)
```
