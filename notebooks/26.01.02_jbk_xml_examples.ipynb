```python
import poolparty as pp
pp.reset_default_party()

marked_seq = 'TCCCGACT.<mut>GGAAAGCGGGCAGTGAGCACACAGGAAA</mut>.ATTACGG.<bc/>.AGATCGGA'
pool = pp.from_seq(marked_seq)\
      .apply_at_marker('bc', transform_fn=lambda _:pp.get_kmers(length=5))\
      .apply_at_marker('mut', transform_fn=lambda _:pp.mutagenize(_, mutation_rate=0.10, mark_changes=True))
df = pool.generate_library(num_seqs=10)
for seq in df['seq']:
    print(seq)
```

```python

import poolparty as pp
pp.reset_default_party()

template = 'TCCCGACT.<mut>GGAAAGCGGGCAGTGAGCACACAGGAAA</mut>.ATTACGG.<bc/>.AGATCGGA'
pool = pp.from_seq(template)
pool = pp.apply_at_marker(pool, 
                          'mut', 
                          transform_fn=lambda _:pp.mutagenize(_, 
                                                              mutation_rate=0.20, 
                                                              mark_changes=True,
                                                              mode='hybrid',
                                                              num_hybrid_states=3))
pool = pp.repeat(pool,times=4)
pool = pp.apply_at_marker(pool, 'bc', transform_fn=lambda _:pp.get_kmers(length=5))
df = pool.generate_library(num_cycles=1)
for seq in df['seq']:
    print(seq)
```

```python
import poolparty as pp
pp.reset_default_party()

template1 = 'TCCCGACT.<mut>GGAAAGCGGGCAGTGAGCACACAGGAAA</mut>.ATTACGG.<bc/>.AGATCGGA'
template2 = 'TCCCGACT.<mut>GGAAAGCAGCACACAGGAAA</mut>.ATTACGG.<bc/>.AGATCGGA'
pool = pp.from_seqs([template1,template2])
pool = pp.apply_at_marker(pool, 
                          'mut', 
                          transform_fn=lambda _:pp.mutagenize(_, 
                                                              mutation_rate=0.20, 
                                                              mark_changes=True,
                                                              mode='hybrid',
                                                              num_hybrid_states=3))
pool = pp.repeat(pool,times=4)
pool = pp.apply_at_marker(pool, 'bc', transform_fn=lambda _:pp.get_kmers(length=5))
df = pool.generate_library(num_cycles=1)
for seq in df['seq']:
    print(seq)
```
