```python
import poolparty as pp
from poolparty import insertion_multiscan
```

```python
# Basic usage: insert 3-char sequences at 2 random positions with a single Pool (deepcopied)
with pp.Party() as party:
    bg = pp.from_seqs(['ACGTACGTACGTACGTACGT'])  # 20 chars
    ins = pp.from_seq('GGG')  # 3-char insertion
    result = insertion_multiscan(bg, num_insertions=2, insertion_pools=ins, spacer_str='.').named('result')
result.generate_seqs(num_seqs=5, seed=42)['seq']
```

```python
# Using multiple distinct pools (different lengths) for insertions
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    pool1 = pp.from_seq('GGG')
    pool2 = pp.from_seq('TTTT')
    result = insertion_multiscan(
        bg, num_insertions=2, insertion_pools=[pool1, pool2], spacer_str='.', 
    ).named('result')
result.generate_seqs(num_seqs=5, seed=42)['seq']
```

```python
# Using hybrid mode with fixed number of states
with pp.Party() as party:
    bg = pp.from_seqs(['ACGTACGTACGTACGTACGT'])  # 20 chars
    ins = pp.from_seq('XXX')
    result = insertion_multiscan(
        bg, num_insertions=2, insertion_pools=ins,
        mode='hybrid', num_hybrid_states=5, spacer_str='.'
    ).named('result')
result.generate_seqs(num_seqs=10, seed=42)['seq']
```

```python
# Three insertions with mark_changes=True (swapcase)
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    ins = pp.from_seq('GGG')
    result = insertion_multiscan(
        bg, num_insertions=3, insertion_pools=ins, mark_changes=True
    ).named('result')
result.generate_seqs(num_seqs=5, seed=42)['seq']
```

```python
# Using spacer_str to mark insertion boundaries
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    ins = pp.from_seq('G')  # 1-char insertion
    result = insertion_multiscan(
        bg, num_insertions=2, insertion_pools=ins, spacer_str='.'
    ).named('result')
result.generate_seqs(num_seqs=5, seed=42)['seq']
```

```python
# insertion_mode='ordered' (default): pool1 always goes to leftmost position, pool2 to next, etc.
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    pool1 = pp.from_seq('GGG')  # Always appears at leftmost selected position
    pool2 = pp.from_seq('TTT')  # Always appears at rightmost selected position
    result = insertion_multiscan(
        bg, num_insertions=2, insertion_pools=[pool1, pool2],
        spacer_str='.', insertion_mode='ordered'
    ).named('result')
df = result.generate_seqs(num_seqs=5, seed=42)
print("insertion_mode='ordered': GGG always left of TTT")
df['seq']
```

```python
# insertion_mode='unordered': pools are randomly assigned to positions
with pp.Party() as party:
    bg = pp.from_seqs(['AAAAAAAAAAAAAAAAAA'])  # 18 chars
    pool1 = pp.from_seq('GGG')  # Can appear at either position
    pool2 = pp.from_seq('TTT')  # Can appear at either position
    result = insertion_multiscan(
        bg, num_insertions=2, insertion_pools=[pool1, pool2],
        spacer_str='.', insertion_mode='unordered'
    ).named('result')
df = result.generate_seqs(num_seqs=10, seed=42)
print("insertion_mode='unordered': GGG and TTT can appear in either order")
df['seq']
```
