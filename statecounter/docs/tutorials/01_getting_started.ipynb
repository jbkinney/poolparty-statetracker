# Tutorial 1: Getting Started with StateCounter

## Basic Usage

All counters must be created within a `Manager` context. Let's start with a simple counter:

````python
from statecounter import Counter, Manager

with Manager():
    A = Counter(num_states=3, name='A')
    print(f"States: {list(A)}")
````

## Combining Counters with Product

Use `product` to create a Cartesian product of counters:

````python
from statecounter import Counter, Manager, product

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = product([A, B], name='C')
    
    print(f"C has {C.num_states} states")
    for state in C:
        print(f"State {state}: A={A.state}, B={B.state}")
````

## Stack: Disjoint Union

Use `stack` to create a disjoint union where only one counter is active at a time:

````python
from statecounter import Counter, Manager, stack

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = stack([A, B], name='C')
    
    print(f"C has {C.num_states} states")
    for state in C:
        if A.is_active():
            print(f"State {state}: A active, A.state={A.state}")
        else:
            print(f"State {state}: B active, B.state={B.state}")
````

## State Propagation

Setting a child counter's state automatically updates all parent counters:

````python
from statecounter import Counter, Manager, product

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = product([A, B], name='C')
    
    C.state = 5  # Set child state
    print(f"After setting C.state = 5:")
    print(f"A.state = {A.state}")
    print(f"B.state = {B.state}")
````

## Visualizing the Counter DAG

Use `print_dag()` to visualize counter dependencies:

````python
from statecounter import Counter, Manager, product

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = product([A, B], name='C')
    
    C.print_dag()
````
