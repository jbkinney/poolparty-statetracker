# Quick Start Guide

This notebook provides a quick introduction to StateCounter, a Python library for creating composable counters with unidirectional state propagation.


## Basic Usage

All counters must be created within a `Manager` context:

````python
from statecounter import Counter, Manager

with Manager():
    A = Counter(num_states=3, name='A')
    print(list(A))
````

## Counter Operations

### Product (Cartesian Product)

Use `product` to create Cartesian products:

````python
from statecounter import Counter, Manager, product

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = product([A, B])  # 6 states
    
    for _ in C:
        print(f"A={A.state}, B={B.state}")
````

### Stack (Disjoint Union)

Use `stack` to create disjoint unions where only one counter is active at a time:

````python
from statecounter import Counter, Manager, stack

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = stack([A, B])  # 5 states
    
    for _ in C:
        # Only one of A or B is active at a time
        print(f"A={A.state}, B={B.state}")
````

### Slicing

Use Python slice syntax to select subsets of states:

````python
from statecounter import Counter, Manager

with Manager():
    A = Counter(num_states=10, name='A')
    B = A[2:5]  # States 2, 3, 4
    C = A[::-1]  # Reversed: 9, 8, 7, ..., 0
    
    print(f"A has {A.num_states} states")
    print(f"B (slice [2:5]) has {B.num_states} states: {list(B)}")
    print(f"C (reversed) has {C.num_states} states: {list(C)}")
````

## State Propagation

StateCounter uses unidirectional state propagation. Setting a child counter's state automatically updates all parent counters:

````python
from statecounter import Counter, Manager, product

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = product([A, B])
    
    C.state = 5  # Set child state
    print(f"After setting C.state = 5:")
    print(f"A.state = {A.state}")  # 1 (automatically updated)
    print(f"B.state = {B.state}")  # 2 (automatically updated)
````

## Visualization

Use `print_dag()` to visualize counter dependencies:

````python
from statecounter import Counter, Manager, product

with Manager():
    A = Counter(num_states=2, name='A')
    B = Counter(num_states=3, name='B')
    C = product([A, B], name='C')
    
    C.print_dag()
````
