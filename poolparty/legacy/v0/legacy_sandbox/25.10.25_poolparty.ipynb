```python
from typing import Tuple, List, Union
    
class Oligo:
    def __init__(self, seq: str, origin=0, parents: Tuple[Union['Oligo', str, int],...] = (), 
                 operation: str = ''):
        self.seq = seq
        self.parents = parents
        self.operation = operation
        self.origin = origin
        
    def visualize_graph(self, indent=0):
        """Print the computation graph structure."""
        print("  " * indent + f"{self.operation or 'input'}: {repr(self)}")
        for parent in self.parents:
            if isinstance(parent, Oligo):
                parent.visualize_graph(indent + 1)
            else: 
                print("  " * (indent + 1) + f"input: {repr(parent)}")

    def __add__(self, other: 'Oligo') -> 'Oligo':
        if isinstance(other, str):
            return Oligo(self.seq + other, self.origin, (self, other), '+')
        else:
            return Oligo(self.seq + other.seq, self.origin, (self, other), '+')
    
    def __radd__(self, other: 'Oligo') -> 'Oligo':
        return self.__add__(other)
    
    def __mul__(self, other: int) -> 'Oligo':
        return Oligo(self.seq * other, self.origin, (self,other), '*',)
    
    def __rmul__(self, other: int) -> 'Oligo':
        return self.__mul__(other)
    
    def __str__(self) -> str:
        return self.seq
    
    def __repr__(self) -> str:
        return f"Oligo('{self.seq}',{self.origin})"
    
    def __len__(self) -> int:
        return len(self.seq)
    
    def __next__(self) -> str:
        return self.seq
```

```python
a = Oligo('AT',1)
b = Oligo('CG',3)
c = 3*a + b + 'ZZZZ'

c.visualize_graph()

print(next(c))
print(next(c))
print(next(c))
```
