```python
import poolparty as pp

with pp.Party() as party:
    background = pp.from_seqs(['T'*20])
    left, right = pp.breakpoint_scan(background, num_breakpoints=1, positions=[3,6,9], mode='sequential')
    insert = pp.from_seqs(['AAA'])
    barcode = pp.get_kmers(length=10, mode='random')
    oligo = left + insert + right + barcode
    party.output(oligo, name='oligo')

df = party.generate(num_complete_iterations=1, seed=42)
print(len(df))
df
```

```python
import poolparty as pp

with pp.Party() as party:
    L = 20
    window_size = 5
    background = pp.from_seqs(['T'*L])
    positions = list(range(1, L-window_size, 3))
    left, middle_right = pp.breakpoint_scan(background, num_breakpoints=1, positions=positions, mode='sequential')
    #insert = pp.from_seqs(['A'*window_size])
    middle = middle_right[:window_size]
    right = middle_right[window_size:]
    middle_shuffled = pp.shuffle(middle, k=3, mode='random')
    oligo = left + middle_shuffled + right
    party.output(oligo, name='oligo')

df = party.generate(num_complete_iterations=1, seed=42)
print(len(df))
df
```
