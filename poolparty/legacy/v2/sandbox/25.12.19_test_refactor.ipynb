```python
# Demonstrate MixOp - mixing sequences from multiple pre-generated pools
from poolparty import from_seqs, mutation_scan, get_kmers, from_motif, mix, Operation

# Reset operation counter for clean output
Operation.id_counter = 0

# Create three different pools
pool_A = from_seqs(['AAA', 'CCC', 'GGG'], mode='sequential', name='pool_A')
pool_B = get_kmers(length=4, mode='sequential', name='pool_B')
pool_C = from_iupac(iupac_seq='NNNN', mode='random', name='pool_C')

# Mix sequences from all three pools, generating different counts from each
mixed_pool = mix(
    pools=[pool_A, pool_B, pool_C],
    num_seqs_by_pool=[3, 10, 5],  # 3 from A, 10 from B, 5 from C
    pool_names=['SeqPool', 'KmerPool', 'MotifPool'],
    mode='sequential',  # Will iterate through all 18 mixed sequences
)

# Generate library and show results with source tracking
results_df = mixed_pool.generate_library(num_seqs=18)
print(f'Generated {len(results_df)} sequences from mixed pool')
print(f'Columns: {list(results_df.columns)}')
results_df
```

```python
# Demonstrate FilterOp - filtering sequences from a pool using various criteria
from poolparty import filter, get_kmers, from_seqs, Operation
from poolparty.filters import gc_range_filter, max_homopolymer_filter

# Reset operation counter for clean output
Operation.id_counter = 0

# Create a pool of 8-mers
kmer_pool = get_kmers(length=8, alphabet='dna', name='kmers')

# Filter using built-in filter parameters:
# - GC content between 40-60%
# - No homopolymer runs longer than 2
# - Minimum edit distance of 3 between accepted sequences
filtered_pool = filter(
    kmer_pool,
    num_candidate_seqs=10000,    # Sample 10,000 candidates
    gc_range=(0.4, 0.6),         # 40-60% GC content
    max_homopolymer=2,           # No runs of 3+ identical bases
    min_edit_distance=3,         # At least 3 edits between sequences
    min_num_filtered_seqs=20,    # Require at least 20 passing sequences
    mode='sequential',
    name='filtered_barcodes'
)

print(f"Number of sequences passing all filters: {filtered_pool.operation.num_states}")

# Generate and display results
results_df = filtered_pool.generate_library(num_seqs=10)
print(f"\nSample of filtered sequences:")
print(results_df[['seq']].head(10))

# Verify the filters worked
print("\nVerifying filter constraints on generated sequences:")
for seq in results_df['seq']:
    gc = sum(1 for c in seq if c in 'GC') / len(seq)
    max_run = max(len(list(g)) for _, g in __import__('itertools').groupby(seq))
    print(f"  {seq}: GC={gc:.0%}, max_homopolymer={max_run}")
```

```python
# Demonstrate FilterOp with custom filter functions
from poolparty import filter, get_kmers, Operation

# Reset operation counter
Operation.id_counter = 0

# Create a pool of 6-mers  
kmer_pool = get_kmers(length=6, alphabet='dna', name='kmers')

# Define custom filter functions
def no_start_codon(seq, filtered_seqs=None):
    """Reject sequences containing ATG (start codon)."""
    return 'ATG' not in seq

def no_stop_codons(seq, filtered_seqs=None):
    """Reject sequences containing stop codons."""
    stop_codons = ['TAA', 'TAG', 'TGA']
    return not any(stop in seq for stop in stop_codons)

# Combine custom filters with built-in filters
filtered_pool = filter(
    kmer_pool,
    num_candidate_seqs=5000,
    filter_funcs=[no_start_codon, no_stop_codons],  # Custom filters
    gc_range=(0.4, 0.6),                             # Built-in: 40-60% GC
    max_homopolymer=2,                               # Built-in: max run of 2
    min_num_filtered_seqs=50,
    mode='sequential',
    name='codon_safe_barcodes'
)

print(f"Sequences passing all filters: {filtered_pool.operation.num_states}")
results_df = filtered_pool.generate_library(num_seqs=10)
print("\nFiltered sequences (no ATG, no stop codons, balanced GC, no homopolymers):")
for seq in results_df['seq']:
    print(f"  {seq}")
```

```python
# Demonstrate multi-length k-mer generation
from poolparty import get_kmers, Operation

# Reset operation counter
Operation.id_counter = 0

# ============================================================================
# SEQUENTIAL MODE: Enumerate all k-mers for each length in order
# ============================================================================
print("="*60)
print("SEQUENTIAL MODE: Multi-length k-mer enumeration")
print("="*60)

# Create pool with lengths 1 and 2 using a small alphabet for clarity
multi_pool = get_kmers(length=[1, 2], alphabet=['A', 'B'], mode='sequential')
print(f"Total states (2^1 + 2^2 = 6): {multi_pool.operation.num_states}")
print(f"Sequence length (variable): {multi_pool.seq_length}")

results_df = multi_pool.generate_library(num_complete_iterations=1)
print("\nAll k-mers enumerated in order:")
print(results_df[['seq', 'op0(get_kmers).length', 'op0(get_kmers).state']])

# ============================================================================
# RANDOM MODE with uniform probabilities
# ============================================================================
print("\n" + "="*60)
print("RANDOM MODE: Multi-length k-mers with uniform probabilities")
print("="*60)

Operation.id_counter = 0
random_pool = get_kmers(length=[3, 6, 9], alphabet='dna', mode='random')
results_df = random_pool.generate_library(num_seqs=12, seed=42)

print("Sample of random k-mers with varying lengths:")
print(results_df[['seq', 'op0(get_kmers).length']])

# Show length distribution
length_counts = results_df['op0(get_kmers).length'].value_counts().sort_index()
print(f"\nLength distribution: {dict(length_counts)}")

# ============================================================================
# RANDOM MODE with custom probabilities (length_probs)
# ============================================================================
print("\n" + "="*60)
print("RANDOM MODE: Biased towards shorter k-mers")
print("="*60)

Operation.id_counter = 0
biased_pool = get_kmers(
    length=[4, 8, 16], 
    alphabet='dna', 
    mode='random',
    length_probs=[0.7, 0.2, 0.1]  # 70% length-4, 20% length-8, 10% length-16
)
results_df = biased_pool.generate_library(num_seqs=100, seed=123)

length_counts = results_df['op0(get_kmers).length'].value_counts().sort_index()
print(f"Length distribution over 100 sequences:")
print(f"  Length 4:  {length_counts.get(4, 0)} sequences (expected ~70)")
print(f"  Length 8:  {length_counts.get(8, 0)} sequences (expected ~20)")
print(f"  Length 16: {length_counts.get(16, 0)} sequences (expected ~10)")
```

```python
# Demonstrate SubseqScanOp - extracting subsequences at specified positions
from poolparty import subseq_scan, from_seqs, Operation

# Reset operation counter
Operation.id_counter = 0

# ============================================================================
# RANGE-BASED INTERFACE: Scan all overlapping windows
# ============================================================================
print("="*60)
print("RANGE-BASED INTERFACE: Scan overlapping windows")
print("="*60)

# Create a pool that extracts all 4-mer windows from a sequence
scan_pool = subseq_scan('ACGTACGTACGT', subseq_length=4, mode='sequential')
print(f"Number of positions: {scan_pool.operation.num_states}")
print(f"Output sequence length: {scan_pool.seq_length}")

results_df = scan_pool.generate_library(num_complete_iterations=1)
print("\nAll 4-mer windows:")
pos_col = [c for c in results_df.columns if 'position' in c][0]
len_col = [c for c in results_df.columns if 'subseq_length' in c][0]
print(results_df[['seq', pos_col, len_col]])

# ============================================================================
# RANGE-BASED with step_size: Non-overlapping windows
# ============================================================================
print("\n" + "="*60)
print("RANGE-BASED: Non-overlapping windows (step_size=4)")
print("="*60)

Operation.id_counter = 0
tiled_pool = subseq_scan('ACGTACGTACGT', subseq_length=4, step_size=4, mode='sequential')
print(f"Number of tiles: {tiled_pool.operation.num_states}")

results_df = tiled_pool.generate_library(num_complete_iterations=1)
print("\nNon-overlapping 4-mer tiles:")
pos_col = [c for c in results_df.columns if 'position' in c][0]
print(results_df[['seq', pos_col]])

# ============================================================================
# POSITION-BASED INTERFACE: Explicit positions
# ============================================================================
print("\n" + "="*60)
print("POSITION-BASED INTERFACE: Specific positions")
print("="*60)

Operation.id_counter = 0
# Extract subsequences only at specific positions of interest
explicit_pool = subseq_scan(
    'ATGCCCGGGAAATTT', 
    subseq_length=3, 
    positions=[0, 3, 6, 9, 12],  # Extract codons at specific positions
    mode='sequential'
)
print(f"Number of positions: {explicit_pool.operation.num_states}")

results_df = explicit_pool.generate_library(num_complete_iterations=1)
print("\nSubsequences at specified positions:")
pos_col = [c for c in results_df.columns if 'position' in c][0]
print(results_df[['seq', pos_col]])

# ============================================================================
# WEIGHTED RANDOM SAMPLING
# ============================================================================
print("\n" + "="*60)
print("WEIGHTED RANDOM SAMPLING: Biased position selection")
print("="*60)

Operation.id_counter = 0
# Create weighted sampling where position 0 is heavily favored
weighted_pool = subseq_scan(
    'ABCDEFGHIJ',
    subseq_length=2,
    positions=[0, 2, 4, 6, 8],
    position_probs=[0.8, 0.05, 0.05, 0.05, 0.05],  # 80% position 0
    mode='random'
)

results_df = weighted_pool.generate_library(num_seqs=20, seed=42)
pos_col = [c for c in results_df.columns if 'position' in c][0]
position_counts = results_df[pos_col].value_counts().sort_index()
print(f"Position distribution over 20 samples:")
for pos, count in position_counts.items():
    print(f"  Position {pos}: {count} samples")

# ============================================================================
# COMBINED WITH OTHER OPERATIONS
# ============================================================================
print("\n" + "="*60)
print("COMBINED: Subsequence scan + mutation + concatenation")
print("="*60)

Operation.id_counter = 0
from poolparty import mutation_scan, get_kmers

# Create a sequence pool
parent = from_seqs(['ACGTACGTACGT', 'TGCATGCATGCA'], mode='sequential', name='parent')

# Extract 6-mer windows, then mutate them
windows = subseq_scan(parent, subseq_length=6, step_size=3, mode='sequential', name='windows')
mutated = mutation_scan(windows, num_mutations=1, mode='random', name='mutated')

# Add a barcode
barcode = get_kmers(length=4, mode='sequential', name='barcode')
final = mutated + '-' + barcode

results_df = final.generate_library(num_seqs=8, seed=42)
print(f"Generated {len(results_df)} sequences")
# Find the relevant columns dynamically
pos_col = [c for c in results_df.columns if 'windows' in c and 'position' in c][0]
mut_col = [c for c in results_df.columns if 'mutated' in c and 'positions' in c][0]
print(results_df[['seq', pos_col, mut_col]].head(8))
```

```python
# Demonstrate multi-length SubseqScanOp - variable-length subsequence extraction
from poolparty import subseq_scan, from_seqs, Operation

# Reset operation counter
Operation.id_counter = 0

# ============================================================================
# MULTI-LENGTH SEQUENTIAL MODE: All positions for each length in turn
# ============================================================================
print("="*60)
print("MULTI-LENGTH SEQUENTIAL: Enumerate by length, then position")
print("="*60)

# Create pool with multiple subsequence lengths
multi_len_pool = subseq_scan(
    'ABCDEFGH',  # 8-char sequence
    subseq_length=[2, 4],  # Extract 2-mers and 4-mers
    mode='sequential'
)
print(f"Total states: {multi_len_pool.operation.num_states}")
print(f"Sequence length (variable): {multi_len_pool.seq_length}")

results_df = multi_len_pool.generate_library(num_complete_iterations=1)
print("\nAll subsequences (length 2 first, then length 4):")
pos_col = [c for c in results_df.columns if 'position' in c][0]
len_col = [c for c in results_df.columns if 'subseq_length' in c][0]
print(results_df[['seq', len_col, pos_col]])

# ============================================================================
# MULTI-LENGTH WITH WEIGHTED PROBABILITIES
# ============================================================================
print("\n" + "="*60)
print("MULTI-LENGTH RANDOM: Biased towards longer subsequences")
print("="*60)

Operation.id_counter = 0
weighted_len_pool = subseq_scan(
    'ACGTACGTACGTACGT',  # 16-char sequence
    subseq_length=[3, 6, 9],
    subseq_length_probs=[0.1, 0.3, 0.6],  # 60% length-9, 30% length-6, 10% length-3
    mode='random'
)

results_df = weighted_len_pool.generate_library(num_seqs=30, seed=42)
len_col = [c for c in results_df.columns if 'subseq_length' in c][0]
length_counts = results_df[len_col].value_counts().sort_index()
print(f"Length distribution over 30 samples:")
for length, count in length_counts.items():
    expected = {3: 3, 6: 9, 9: 18}[length]
    print(f"  Length {length}: {count} samples (expected ~{expected})")

# ============================================================================
# MULTI-LENGTH WITH EXPLICIT POSITIONS
# ============================================================================
print("\n" + "="*60)
print("MULTI-LENGTH WITH POSITIONS: Different valid ranges per length")
print("="*60)

Operation.id_counter = 0
# Position 6 is valid for length 2 but not for length 4 (would exceed string)
multi_pos_pool = subseq_scan(
    'ABCDEFGH',  # 8-char sequence
    subseq_length=[2, 4],
    positions=[0, 2, 4, 6],  # Position 6 only valid for length 2
    mode='sequential'
)

print(f"Total states: {multi_pos_pool.operation.num_states}")
print("  - Length 2: positions 0,2,4,6 (4 states)")
print("  - Length 4: positions 0,2,4 only (3 states, position 6 would overflow)")

results_df = multi_pos_pool.generate_library(num_complete_iterations=1)
print("\nSubsequences with position filtering:")
pos_col = [c for c in results_df.columns if 'position' in c][0]
len_col = [c for c in results_df.columns if 'subseq_length' in c][0]
print(results_df[['seq', len_col, pos_col]])
```
